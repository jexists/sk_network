## **9.1.apply 함수**
- 사용자 정의 함수를 데이터에 적용하고 싶을 때 사용한다.
- `데이터프레임.apply(함수, axis=0/1)`

````python
df1 = df.copy()
````

````python
# 성인인지 아닌지 (19세 이상은 성인)

def adult(x):
  if x >= 19:
    return 1
  elif x < 19:
    return 0
  else:
    return np.nan


# def adult(x):
#   if x >= 19:
#     return 1
#   return 0
````

````python
df1['adult'] = df1['Age'].apply(adult)
df1.head(10)
````

- 간단한 함수는 lambda로 구현할 수 있다.

````python
# Pclass가 1등급이고, SipSp가 1인 경우 1. 그렇지 않으면 0.
df1['pclass_sibsp_lamda'] = df1.apply(lambda x: 1 if x['Pclass'] == 1 and x['SibSp'] == 1 else 0, axis=1)

df1.head()
````

## **9.2.문자열 다루기**

|메소드|설명|
|:--|:--|
|`.str.contains(문자열)`|문자열을 포함하고 있는지 유무|
|`.str.replace(기존문자열, 대치문자열)`|문자열 대치|
|`.str.split(문자열, expand=True/False, n=개수)`|특정 문자열을 기준으로 쪼개기|
|`.str.lower()`|소문자로 바꾸기|
|`.str.upper()`|대문자로 바꾸기|

````python
df2 = df.copy()
df2.head()
````

````python
df2['Name'].str.contains('Miss')
````

````python
df2[df2['Name'].str.contains('Miss')]
````

````python
df2['Name'] = df2['Name'].str.replace(',','')
df2.head()
````

````python
df2['Name'].str.split(' ')
````

- `expand=False` 가 기본값이므로 리스트 형태로 반환

````python
df2['Name'].str.split(' ', expand=True, n=1)
# n=1은 최대 1번만 나누기
````

````python
# 소문자로 바꾸기
df2['Name'].str.lower()
````

````python
# 대문자로 바꾸기
df2['Name'].str.upper()
````

## **9.3.피벗테이블**
- 엑셀의 피벗테이블처럼 인덱스별 컬럼별 값의 연산을 할 수 있다.
- `pd.pivot_table(데이터명, index=, columns=, values=, aggfunc=)`


- 기본 문법
```
pd.pivot_table(data,
               index=인덱스로 사용할 열,
               columns=열로 사용할 열,
               values=집계할 값,
               aggfunc=집계 함수,
               margins=True) # 전체 행과 열의 합/평균 추가
```



````python
df3 = df.copy()
df3.head()
````

**1) 단일 인덱스, 단일 컬럼, 단일 값**

````python
pd.pivot_table(
                df3,
                index='Sex',
                columns='Pclass',
                values='Survived',
                aggfunc='mean'
               )
````

````python
pd.pivot_table(
                df3,
                index='Pclass',
                columns='Sex',
                values='Survived',
                aggfunc='mean'
               )
````

````python
pd.pivot_table(
                df3,
                index='Sex',
                columns='SibSp',
                values='Survived',
                aggfunc='mean'
               )
````

- `margins` 옵션을 통해 행과 열 전체의 값도 구할 수 있음

````python
pd.pivot_table(
                df3,
                index='Sex', # 인덱스로 사용할 열
                columns='Survived', #열로 사용할 열
                values='Age', # 집계할 값
                aggfunc='mean', # 집계 함수
                margins=True
               )
````

````python
pd.pivot_table(
                df3,
                index='Sex',
                columns='Survived',
                values='Age',
                aggfunc=['mean', 'max', 'min']
               )
````

````python
pd.pivot_table(
                df3,
                index='Sex',
                columns='Survived',
                values='Age',
                aggfunc=['mean', 'max', 'min'],
                margins=True
               )
````

**2) 다중 인덱스, 다중 컬럼, 다중 값**

````python
pd.pivot_table(
                df3,
                index=['Sex','Pclass'],
                columns='Survived',
                values='Age',
                aggfunc='mean'
               )
````

````python
pd.pivot_table(
                df3,
                index=['Sex','Pclass'],
                columns=['Survived', 'Embarked'],
                values='Age',
                aggfunc='mean'
               )
````

**3) pivot_table VS groupby**

| **기능**       | **`pd.pivot_table`**       | **`groupby`**           |
|---------------|---------------------------|-------------------------|
| **목적**      | 행, 열, 값으로 다차원 분석       | 주로 그룹화 및 집계        |
| **출력 형태**   | 피벗 테이블 형태의 `DataFrame` | 그룹화된 `DataFrame` 또는 `Series` |
| **다중 집계 함수** | 지원 (`aggfunc`로 여러 함수 가능) | `agg()`로 여러 함수 가능   |
| **margins 옵션** | 전체 합/평균 등 추가 가능       | 지원하지 않음             |
| **다중 인덱스**   | 행과 열에 다중 인덱스 생성       | 행에만 다중 인덱스 생성      |
| **사용성**     | 행과 열을 동시에 요약 가능       | 주로 행 기준으로 요약      |
