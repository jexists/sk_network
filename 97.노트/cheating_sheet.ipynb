### 데이터 이해

```python
df.head() # 상위 행 확인
df.tail() # 하위 행 확인

df.info() # 변수 확인
df.describe() # 기술통계 확인 (숫자 컬럼만)
# count	O
# mean	O
# std	O
# min/25%/50%/75%/max	O
df.describe(include='object')    # 문자열 컬럼만
# count	비어있지 않은 값 개수
# unique	고유값 개수
# top	가장 많이 나온 값
# freq	top 값의 등장 횟수
df.describe(include='all')       # 전체 (수치+범주, 해당없는 칸은 NaN)


df.corr() # 상관 관계 확인
df.corr(numeric_only=True)

df.isna().sum() # 결측치 처리
df.isnull().sum()

df.duplicated().sum() # 중복 데이터 확인

df[''].unique() # 열의 고유값 확인

df[''].value_counts()
```
