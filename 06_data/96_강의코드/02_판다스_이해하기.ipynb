## 1.1.판다스(Pandas)란?
- 빠르고 유연한 데이터 구조를 제공하는 Python 라이브러리
- 행과 열로 이루어진 2차원 데이터를 효율적으로 가공


### 1.1.1. 판다스를 사용하는 이유
- 다양한 데이터 처리 기능 제공
    - 데이터 읽고 쓰기, 결측치 처리, 그룹화, 필터링, 병합, 집계 등 다양한 데이터 처리 기능 제공
- 유연한 인덱싱
    - 다양한 인덱싱 방법을 제공하여 데이터프레임에서 데이터를 편리하게 선택하고 조작 가능
- 통계 및 요약 정보 제공
    - 데이터에 대한 통계적 분석과 요약 정보를 제공하여 데이터의 특성을 빠르게 파악할 수 있도록 도와줌
- 시각화 지원
    - 다양한 시각화 라이브러리와 함께 사용되어 데이터를 시각화하는데 유용한 기능을 제공

## 1.5.CSV파일 읽어오기

- 분석용 데이터는 대부분 파일에서 읽어 가져오니 잘 익혀야 할 기능이다.
- **read_csv()** 함수를 사용해서 CSV 파일에서 데이터를 읽어온다.

**[csv 파일이란?]**
- csv 파일은 데이터를 텍스트와 쉼표로 구분하여 저장하는 파일 형식이다.
- 파일이 텍스트와 쉼표만으로 이루어져 있어 구조가 매우 간단하고 용량이 작은 편이다.

**[주요옵션]**
- sep: 구분자 지정(기본값 = 콤마)
- header: 헤더가 될 행 번호 지정(기본값 = 0)
- index_col: 인덱스 열 지정(기본값 = False)
- names: 열 이름으로 사용할 문자열 리스트
- encoding: 인코딩 방식을 지정
