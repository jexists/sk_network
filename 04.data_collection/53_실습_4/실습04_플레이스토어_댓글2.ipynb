```python
# 라이브러리 불러오기
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
# 2. 드라이버 옵션 설정 및 브라우저 띄우기
driver = webdriver.Chrome()
```

```python
# 3. 주소 이동(플레이스토어 첫 화면)
url = "https://play.google.com/store/games?hl=ko-KR"
driver.get(url)
time.sleep(2)
```

```python
# 4. 실제 앱 주소로 이동

app_package = "com.devsisters.CookieRunForKakao"
app_url = f"https://play.google.com/store/apps/details?id={app_package}&hl=ko-KR"
driver.get(app_url)
```

```python
# 5-1. 댓글 수집
comment_list = []

list = driver.find_elements(By.CSS_SELECTOR, '.h3YV2d')

print(len(list))

# print(list[0].text.strip())
```

```python
# 5-2. '리뷰 모두 보기' 버튼 클릭
# copy_selector = '#ow8 > section > div > div.Jwxk6d > div:nth-child(5) > div > div > button > span'
copy_selector = '#yDmH0d > c-wiz.SSPGKf.Czez9d > div > div > div:nth-child(1) > div > div.wkMJlb.YWi3ub > div > div.qZmL0 > div:nth-child(1) > c-wiz:nth-child(6) > section > header > div > div:nth-child(2) > button'

try:
    review_button = driver.find_element(By.CSS_SELECTOR, copy_selector)
    review_button.click()
    time.sleep(2)
except Exception as e:
    print("리뷰 모두 보기 버튼 클릭 실패:", e)
```

```python
# 6-1. 리뷰 확인

reviews = driver.find_elements(By.CSS_SELECTOR, "div.h3YV2d")
print(len(reviews))
```

```python
# 6-2. 스크롤 다운 후 리뷰 확인
page = driver.find_element(By.CLASS_NAME, 'fysCi')
page.click()        #  팝업창 클릭
driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
reviews = driver.find_elements(By.CSS_SELECTOR, "div.h3YV2d")
print(len(reviews))
time.sleep(2)
```

```python
# 6-3. while 문으로 스크롤 다운 작성
target_review = 200
while True:
    reviews = driver.find_elements(By.CSS_SELECTOR, "div.h3YV2d")
    print(len(reviews), "만큼 수집하였습니다.")
    if len(reviews) < target_review:  # 목표 리뷰 숫자보다 작을 경우
        # 스크롤 다운
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(1)
    else:
        break
```

```python
# 7-1. 리스트로 저장
result_list = []
for review in reviews:
    result_list.append(review.text)
result_list
```

```python
# 7-2. 맨 앞 3개의 리뷰 중복으로 인해 제거
final_review_list = result_list[3:]
final_review_list[:20]      # 20개만 출력
```
