## 1.웹 스크래핑 방법

#### 웹페이지의 종류
- 정적인 페이지 : 웹 브라우져에 화면이 한번 뜨면 이벤트에 의한 화면의 변경이 없는 페이지
- 동적인 페이지 : 웹 브라우져에 화면이 뜨고 이벤트가 발생하면 서버에서 데이터를 가져와 화면을 변경하는 페이지

#### requests 이용
- 받아오는 문자열에 따라 두가지 방법으로 구분
    - json 문자열로 받아서 파싱하는 방법 : 주로 동적 페이지 크롤링할때 사용
    - html 문자열로 받아서 파싱하는 방법 : 주로 정적 페이지 크롤링할때 사용
        
#### selenium 이용
- 브라우져를 직접 열어서 데이터를 받는 방법

#### 크롤링 방법에 따른 속도
- requests json > requests html > selenium

## 3.네이버 날씨

### 3.1. 라이브러리 불러오기

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

### 3.2. html 요소 얻기

```python
keyword = '서울'

url = f'https://search.naver.com/search.naver?where=nv&sm=top_sug.pre&fbm=0&acr=4&acq=%EC%84%9C%EC%9A%B8+%EB%82%A0%EC%94%A8&qdt=0&ie=utf8&query={keyword}+날씨'

page = requests.get(url)

page
```

### 3.3. BeatifulSoup 으로 파싱

```python
bs_result = BeautifulSoup(page.text, 'html.parser')

bs_result
```

### 3.4. 기온 요소 추출

```python
# 방법 1 ) 두번에 걸쳐서 가져오기
today_temperature = bs_result.find('div', class_='_today')

data = today_temperature.find('div', class_='temperature_text')

print(today_temperature)

print(data)
```

```python
# data 객체 (보통 BeautifulSoup로 찾은 태그)안에 텍스트만 추출
# separator : 태그 사이에 구분자를 넣지 않고 글자를 그대로 이어붙이겠다.
# strip : 앞뒤 공백과 줄바꿈을 제거
now_temperature = data.get_text(separator="", strip=True)
now_temperature
```

```python
# 방법2) 한번에 가져오기
temperature = bs_result.find_all('div', class_='temperature_text')
temperature
```

```python
now_temperature2 = temperature[0].get_text(separator='')
now_temperature2
```

```python
# 방법 3 copy selector를 이용해서 가져오기

result_list = []

# 현재 기온이 들어있는 strong 태그를 정확한 css 경로로 지정
# copy < copy selector
copy_selector = "#main_pack > section > div > div:nth-child(1) > div.content_wrap > div.open > div:nth-child(1) > div > div.weather_info > div > div._today > div.weather_graphic > div.temperature_text > strong"

# 선택자에 해당하는 첫 번째 요소 1개를 가져옴
copy_data = bs_result.select_one(copy_selector)

print(copy_data)
```

```python

# 가져온 요소에서 텍스트만 추출
copy_text = copy_data.get_text(separator='')

print(copy_text)

result_list.append(copy_text)

print(result_list)
```
