# 3.단변량 분석 : 숫자형 변수

```python
def eda_num(df, col, bins=20):
    # 기초통계량
    display(df[col].describe().to_frame().T)
    print(f'왜도(skewness): {df[col].skew():.4f}  |  첨도(kurtosis): {df[col].kurt():.4f}')

    # 시각화
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    df[col].hist(ax=axes[0], bins=bins, edgecolor='k', color='steelblue', alpha=0.7)
    axes[0].axvline(df[col].mean(),   color='red',   linestyle='--', label=f'mean={df[col].mean():.2f}')
    axes[0].axvline(df[col].median(), color='green', linestyle='--', label=f'median={df[col].median():.2f}')
    axes[0].set_title(f'{col} Histogram')
    axes[0].legend()

    axes[1].boxplot(df[col], vert=True)
    axes[1].set_title(f'{col} Boxplot')
    axes[1].set_ylabel(col)

    plt.suptitle(col, fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

# 4.단변량 분석 : 범주형 변수

```python
def eda_cat(df, col):
    # 기초통계량
    vc = df[col].value_counts()
    vc_ratio = df[col].value_counts(normalize=True)
    display(pd.DataFrame({'count': vc, 'ratio(%)': vc_ratio * 100}))

    # 시각화
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    vc.plot(kind='bar', ax=axes[0], color='steelblue', edgecolor='k', alpha=0.7)
    axes[0].set_title(f'{col} - 빈도')
    axes[0].set_ylabel('Count')
    axes[0].tick_params(axis='x', rotation=0)
    for i, v in enumerate(vc):
        axes[0].text(i, v + 1, str(v), ha='center', va='bottom')

    axes[1].pie(vc, labels=vc.index, autopct='%1.1f%%', startangle=90)
    axes[1].set_title(f'{col} - 비율')

    plt.suptitle(col, fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.show()
```
