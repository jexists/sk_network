# 다룰 내용

- sklearn에서 제공하는 토이 데이터가 무엇인지, 어떻게 활용하는지 살펴보자.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

- load_iris() 함수로 붓꽃 품종 관련 데이터를 가져올 수 있다.

```python
from sklearn.datasets import load_iris

iris = load_iris()

iris.keys()
```

- DESCR 속성에 데이터에 대한 상세한 설명이 포함되어 있다.

```python
print(iris.DESCR)
```

- feature_names 속성에 독립변수들 이름이 포함되어 있다.
- data 속성에 독립변수들 값이 포함되어 있다.

```python
print(iris.feature_names)
print(iris.data[:10])
```

- target_names 속성에 종속변수 이름이 포함되어 있다.
- target 속성에 종속변수 값이 포함되어 있다.

```python
print(iris.target_names)
print(iris.target[:10])
```

- 위 정보들을 활용해 데이터프레임을 만들 수 있다.

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df.head()
```

```python
# 변수 이름 변경
df.columns = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']

# 확인
df.head()
```

- 데이터프레임에 종속변수, 즉 target이 되는 Species 변수를 추가한다.

```python
df['Species'] = iris.target

df.head()
```

- Species 변수의 값을 품종 이름으로 변경하면 우리가 아는 Iris 데이터프레임이 된다.

```python
df['Species'].replace([0,1,2], iris.target_names, inplace=True)

df.head()
```

- Iris 데이터는 품종별로 50씩의 관측치를 갖는다.

```python
df['Species'].value_counts()
```
