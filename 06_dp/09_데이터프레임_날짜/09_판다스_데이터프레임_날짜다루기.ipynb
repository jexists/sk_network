## **8.1.문자형을 날짜형으로 변경**
- 날짜가 문자형으로 되어있다면 날짜형으로 변경해야 여러가지 날짜 계산을 할 수 있다.
- `pd.to_datetime(컬럼, format='날짜 형식')`

|형식|설명|
|:--|:--|
|%Y|0을 채운 4자리 연도|
|%y|0을 채운 2자리 연도|
|%m|0을 채운 월|
|%d|0을 채운 일|
|%H|0을 채운 시간|
|%M|0을 채운 분|
|%S|0을 채운 초|

## **8.2.날짜를 원하는 형식으로 변경**
`데이터컬럼.dt.strftime(날짜형식)`

## **8.3.dt 연산자**

|연산자|설명|
|:--|:--|
|year|연도|
|month|월|
|day|일|
|dayofweek|요일(0-월요일, 6-일요일)|
|day_name()|요일을 문자열로|

## **8.4.날짜 계산**
- day 연산: `pd.Timedelta(day=숫자)`
- month 연산: `DateOffset(months=숫자)`
- year 연산: `DateOffset(years=숫자)`

- y축에 여러 값을 표시할 수 있다. 이때는 범례를 표시하는 것이 좋다.

## **8.5.기간 이동 계산**
- 시간에 따라 변화하는 데이터의 추세를 파악하기 위해 사용되는 통계 기법

`컬럼.rolling().집계함수`

- 평균 외에 합계, 최솟값, 최댓값 등 다양한 연산이 가능하다.
