## 3.2.특정 열 조회

- **df.loc[ : , [열 이름1, 열 이름2,...]]** 형태로 조회할 열 이름을 리스트로 지정한다.
- 조회할 열이 하나면 리스트 형태가 아니어도 된다.
- 열 부분은 생략할 수 있었지만, **행 부분을 생략할 수는 없다.**
- 하지만 **df[[열 이름1, 열 이름2,...]]** 형태로 인덱서를 생략함이 일반적이다.

```python
# total_bill 열 조회
tips['total_bill']
```

```python
# total_bill 열 조회
tips[['total_bill']]
```

```python
tips.loc[:,['total_bill']]
```

```python
# tip, total_bill 열 조회 (2개)

# 방법 2: loc 사용 시 행은 : 로 전체 선택, 열은 두 번째 인자
tips.loc[:, ['tip', 'total_bill']]
```

## 3.3.열 범위 조회

- 범위 조회는 **df.loc[:, 열 이름1:열 이름2]** 형태로 조회한다.
- 범위는 하나만 지정할 수 있어 리스트가 될 수 없으니 대괄호를 사용하지 않는다.
- 범위 마지막 열도 조회 대상에 포함된다.


```python
# tips.head()

# sex ~ time 열 조회
tips.loc[:, 'sex':'time']
```

## 3.4.조건으로 조회

- **df.loc[조건]** 형태로 조건을 지정해 조건에 만족하는 데이터만 조회할 수 있다.

**1) 단일 조건 조회**

```python
# 조건: tips['tip'] > 6.0

tips.loc[tips['tip']>6.0,:]
```

- 조건을 변수로 선언해 사용할 수 있다.

```python
cond = tips['tip']>6.0

tips.loc[cond,:]
```

**2) 여러 조건 조회**

- [ ]안에 조건을 여러개 연결할 때 **and와 or 대신에 &와 |** 를 사용해야 한다.
- 그리고 각 조건들은 **(조건1) & (조건2)** 형태로 **괄호**로 묶어야 한다.

```python
# tip 열 값이 6.0 보다 크고, day가 Sat 조회

con = (tips['tip'] > 6.0) & (tips['day']=='Sat')

tips.loc[cond]
```

```python
# tip 열 값이 6.0 보다 크거나 또는 day가 Sat 조회

con = (tips['tip'] > 6.0) | (tips['day']=='Sat')

tips.loc[cond]
```

**3) 편리한 isin(), between() 메소드**

- **isin([값1, 값2,..., 값n])**: 값1 또는 값2 또는...값n인 데이터만 조회한다.

```python
# 값 나열 (day가 Sat 또는 Sun)
# tips.loc[tips['day'].isin(['Sat','Sun']),:]
tips.loc[tips['day'].isin(['Sat','Sun'])]
```

- 위 구문은 다음과 같은 의미를 갖는다.

```python
tips.loc[(tips['day']=='Sat') | (tips['day']=='Sun')]
```

- **between(값1, 값2)**: 값1 ~ 값2까지 범위안의 데이터만 조회한다.

```python
# 범위 지정 (1 <= size <= 3)
tips.loc[tips['size'].between(1,3)]
# tips.loc[tips['size'].between(1,3),:]
```

- 위 구문은 다음과 같은 의미를 갖는다.

```python
# and
tips.loc[(tips['size']>=1) & (tips['size']<=3)]
```

**4) 조건을 만족하는 행의 일부 열 조회**

- **df.loc[조건, ['열 이름1', '열 이름2',...]]** 형태로 조회할 열을 리스트로 지정한다.

```python
# 조건에 맞는 하나의 열 조회

tips.loc[tips['size'] >= 5, ['tip']]
```

```python
# 조건에 맞는 여러 열 조회


tips.loc[tips['size'] >= 5, ['tip', 'total_bill']]
```

```python
# 가능하지만 비권장
tips.total_bill
```

## 3.5.조회 정리

- **데이터프레임 `loc` 조회 예시**

| **조회**        | **코드 예**                                                                                 |
|------------------|--------------------------------------------------------------------------------------------|
| **특정 열**     | df.loc[:, 'score'] → `df['score']`                                                        |
| **여러 열**     | df.loc[:, ['name', 'score']] → `df[['name', 'score']]`                                    |
| **특정 행**     | df.loc[1, :] → `df.loc[1]`                                                               |
| **여러 행**     | df.loc[[1, 3, 5], :] → `df.loc[[1, 3, 5]]`                                               |
| **특정 조건**   | df.loc[df['score'] >= 80, :] → `df.loc[df['score'] >= 80]`                                |
| **여러 조건**   | df.loc[(df['score'] >= 80) & (df['age'] < 20), :] → `df.loc[(df['score'] >= 80) & (df['age'] < 20)]` |
| **여러 조건 특정 열** | `df.loc[(df['score'] >= 80) & (df['age'] < 20), ['name', 'age', 'score']]`                |
| **between()**   | df.loc[df['score'].between(80, 90), :] → `df.loc[df['score'].between(80, 90)]`            |
| **isin()**      | df.loc[df['grp'].isin(['A1', 'B3', 'C1']), :] → `df.loc[df['grp'].isin(['A1', 'B3', 'C1'])]` |


- 원래 코드가 무엇인지 알고 있어야 한다.
- 조건 조회 시 열을 명시하지 않고 조건에 맞는 행 전체를 선택할 경우 loc를 생략할 수 있다.
- df['score'] 를 df.score로 작성할 수 있으나 df['score'] 형태로 사용하기를 권고한다.

## 3.6.인덱스 초기화

- 다음과 같은 경우 인덱스 초기화가 필요하다.
    - 기존 데이터프레임에서 일부 행을 가져와 새로 만든 데이터프레임
    - 일부 행이 지워진 데이터프레임
- 특히 0부터 시작하는 정수형 인덱스의 경우 초기화를 하는 것이 좋다.
- 깔끔한 데이터가 사용하기 좋은 데이터이다.

```python
# tip 열이 6이상인 행 추출
good = tips.loc[tips['tip']>=6]
good
```

- 기존 인덱스 값은 의미가 없으므로 drop=True 옵션을 지정해 제거한다.
- inplace=True 옵션을 지정하지 않으면 실제 반영되지 않는다.


```python
good.reset_index(drop=True, inplace=True)
```

## 3.7. iloc[ ]로 필터링하기
- iloc는 정수 위치(integer location)의 줄임말로 정수값으로 필터링할 때 사용
- loc[ ]와 비슷하지만 필터링할 때 대상을 정수로 짚어내야 하는 점이 다르다.

**- loc와 iloc 특징 및 차이점**

| **기능**      | **loc**               | **iloc**            |
|---------------|------------------------|---------------------|
| **인덱싱 기준** | **라벨(Label)**        | **정수(Integer)**   |
| **범위 선택**   | **포함(inclusive)**    | **제외(exclusive)** |
| **에러 발생**   | 잘못된 라벨 사용 시 에러 | 잘못된 인덱스 사용 시 에러 |
| **속도**      | 조금 느릴 수 있음       | 상대적으로 빠름     |
| **가독성**    | 매우 좋음             | 다소 떨어질 수 있음  |
| **사용 예시**   | `df.loc['2024-01-01']` | `df.iloc[0]`       |


```python
tips.head()
```

```python
# .iloc[] : 데이터 프레임의 행과 열 번호를 사용
# 3번째, 4번째 행(row)과 1번째, 3번째 열을 출력 (2x2)

result = tips.iloc[[2,3],[0,2]]
print(result)
```
