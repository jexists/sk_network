## 다른 풀이 방식

### 풀이 1 - sort() 인플레이스 (원본 직접 수정)

`sorted()`는 새 리스트를 반환하고, `sort()`는 원본을 직접 수정한다.

```python
def solution(strings, n):
    strings.sort(key=lambda x: (x[n], x))
    return strings

solution(["sun", "bed", "car"], 1)
```

### 풀이 2 - cmp_to_key (비교 함수 직접 정의)

두 값을 직접 비교하는 함수를 작성. 복잡한 정렬 조건에 유용하다.

```python
from functools import cmp_to_key

def solution(strings, n):
    def compare(a, b):
        if a[n] != b[n]:
            return -1 if a[n] < b[n] else 1
        return -1 if a < b else (1 if a > b else 0)

    return sorted(strings, key=cmp_to_key(compare))

solution(["sun", "bed", "car"], 1)
```

### 풀이 3 - 두 번 정렬 (안정 정렬 활용)

파이썬 sort는 **안정 정렬(stable sort)** 이므로, 2차 기준 먼저 정렬한 뒤 1차 기준으로 다시 정렬하면 순서가 유지된다.

```python
def solution(strings, n):
    result = sorted(strings)           # 1단계: 알파벳 전체 순 정렬 (2차 기준)
    result = sorted(result, key=lambda x: x[n])  # 2단계: n번째 문자 기준 (1차 기준)
    return result

solution(["sun", "bed", "car"], 1)
```

### 풀이 4 - 버블 정렬 (직접 구현)

알고리즘 원리 이해용. O(n²)이라 실전에서는 비효율적.

```python
def solution(strings, n):
    arr = strings[:]
    for i in range(len(arr)):
        for j in range(len(arr) - i - 1):
            a, b = arr[j], arr[j+1]
            if (a[n], a) > (b[n], b):
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

solution(["sun", "bed", "car"], 1)
```

### 풀이 5 - sorted 중첩 (한 줄)

`sorted()` 안에 `sorted()`를 넣어 풀이 3을 한 줄로 표현. 안쪽 sorted로 알파벳 정렬 후, 바깥 sorted로 n번째 문자 기준 재정렬.

```python
def solution(strings, n):
    return sorted(sorted(strings), key=lambda x: x[n])

solution(["sun", "bed", "car"], 1)
```
