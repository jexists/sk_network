# 같은 숫자는 싫어

https://school.programmers.co.kr/learn/courses/30/lessons/12906

````python
def solution(arr):
    answer = []
    last = ''
    for i in arr:
        if last != i:
            answer.append(i)
            last = i
    return answer
````

## 내 풀이 리뷰

**접근법**: 이전 값을 `last` 변수에 저장하며 현재 값과 다를 때만 append

**장점**
- 직관적이고 읽기 쉬움
- O(n) 시간복잡도, O(n) 공간복잡도

**개선 포인트**
- `last = ''`로 초기화하면 arr에 빈 문자열이 들어올 경우 첫 원소가 누락될 수 있음
- `last = None`으로 초기화하는 게 더 안전함 (정수 배열이므로 실제로는 문제없지만)

## 다른 사람 풀이 1 — `itertools.groupby` 활용

````python
from itertools import groupby

def solution(arr):
    return [k for k, _ in groupby(arr)]
````

**풀이 설명**

`itertools.groupby`는 연속된 동일 원소를 묶어주는 표준 라이브러리 함수.

```
arr = [1, 1, 3, 3, 0, 1, 1]
groupby(arr) → (1, iter[1,1]), (3, iter[3,3]), (0, iter[0]), (1, iter[1,1])
```

`k`만 추출하면 연속 중복이 제거된 리스트가 됨.  
내 풀이와 동일한 O(n) 복잡도지만 코드가 한 줄로 훨씬 간결함.

## 다른 사람 풀이 2 — `zip`으로 현재/이전 비교

````python
def solution(arr):
    return [a for a, b in zip(arr, [None] + arr) if a != b]
````

**풀이 설명**

`[None] + arr`을 만들어 원본 arr과 zip하면 `(현재값, 이전값)` 쌍을 얻을 수 있음.

```
arr        = [1, 1, 3, 3, 0, 1, 1]
[None]+arr = [None, 1, 1, 3, 3, 0, 1, 1]

zip 결과 →  (1,None), (1,1), (3,1), (3,3), (0,3), (1,0), (1,1)
a != b  →   ✓        ✗      ✓      ✗      ✓      ✓      ✗

결과: [1, 3, 0, 1]
```

첫 원소는 항상 `None`과 비교되므로 무조건 포함됨. 리스트 컴프리헨션 한 줄로 해결하는 트릭성 코드.

**단점**: `[None] + arr`로 새 리스트를 만들어 메모리를 추가 사용함.
