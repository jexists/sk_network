# [PCCE 기출문제] 10번 / 데이터 분석

https://school.programmers.co.kr/learn/courses/30/lessons/250121

```python
def solution(data, ext, val_ext, sort_by):
    # print(data)

    answer = []
    filter = 0
    sort = 0
    if ext == 'code':
        filter = 0
    elif ext == 'date':
        filter = 1
    elif ext == 'maximum':
        filter = 2
    elif ext == 'remain':
        filter = 3

    if sort_by == 'code':
        sort = 0
    elif sort_by == 'date':
        sort = 1
    elif sort_by == 'maximum':
        sort = 2
    elif sort_by == 'remain':
        sort = 3
    for i in data:
        # print(i)
        # print(i[filter])
        if i[filter] <= val_ext:
            answer.append(i)
        
    # answer = sorted(answer, key=lambda x: x[sort],reverse=True)
    
    answer = sorted(answer, key=lambda x: x[sort])
    
    # print(answer)
    return answer


solution(
    [[1, 20300104, 100, 80], [2, 20300804, 847, 37], [3, 20300401, 10, 8]],
    "date",
    20300501,
    "remain"
)
```

```python


def solution(data, ext, val_ext, sort_by):
    header = {'code':0, 'date':1, 'maximum':2, 'remain':3}
    answer = []

    for i in data:
        if i[header[ext]] <= val_ext:
            answer.append(i)
    
    answer = sorted(answer, key=lambda x: x[header[sort_by]])

    return answer


solution(
    [[1, 20300104, 100, 80], [2, 20300804, 847, 37], [3, 20300401, 10, 8]],
    "date",
    20300501,
    "remain"
)
```

## 다른사람 코드

```python

def solution(data, ext, val_ext, sort_by):
    answer = []
    by = [ "code", "date", "maximum", "remain" ]

    for item in data:
        if item[by.index(ext)] < val_ext:
            answer.append(item)

    return sorted(answer, key=lambda x: x[by.index(sort_by)])
```
