## 가장 가까운 같은 글자


https://school.programmers.co.kr/learn/courses/30/lessons/142086

```python
def solution(s):
  # split 공백 나눔 / 기준
  str_list = list(s)
  # print(str_list)

  set_list = {}
  answer = []

  for i, v in enumerate(list(s)):
    # print(v)
    # print(set_list.get(v))
    if (set_list.get(v) == None):
      set_list[v] = i
      answer.append(-1)
    else:
      # print(f'{i} - {set_list[v]}')
      answer.append(i - set_list[v])
      set_list[v] = i

  # print(set_list)
  return answer

print(solution('banana'))
```

```python
# 수정사항
def solution(s):
  str_list = list(s)

  set_list = {}
  answer = []

  for i, v in enumerate(str_list):
    if (set_list.get(v) == None):
      # get(v) == None < not in
      set_list[v] = i
      answer.append(-1)
    else:
      answer.append(i - set_list[v])
      set_list[v] = i
  return answer
```

## 다른사람 코드

```python
# dict() + not in 사용
# dic.get(v) == None 대신 not in 으로 더 간결하게 존재 여부 확인
# enumerate 대신 range(len(s)) 로 인덱스 접근

# dict not in	딕셔너리로 마지막 위치 저장	O(n)
def solution(s):
    answer = []
    dic = dict()
    for i in range(len(s)):
        if s[i] not in dic:
            answer.append(-1)
        else:
            answer.append(i - dic[s[i]])
        dic[s[i]] = i

    return answer
```

```python
# s[i::-1]로 현재 위치부터 역방향 슬라이싱
def solution(s):
    return [s[i::-1].find(s[i],1) for i in range(len(s))]

# 예) "banana", i=3 (두 번째 a) 이면:
# s[3::-1] → "anab" (뒤집기)
# .find('a', 1) → 인덱스 1에서부터 'a' 찾기 → 결과 2

# 뒤집었기 때문에 찾은 인덱스가 곧 거리가 됨. 없으면 find()가 -1 반환하므로 조건문 불필요

# 슬라이싱 find	역방향 슬라이싱	O(n²)
```

```python
# 리스트 + index() + 찾은 글자 덮어쓰기
def solution(s):
    answer = []
    t = list(s)
    for i in range(len(t)):
        if t[i] in t[:i]:
            answer.append(i-t.index(t[i]))
            t[t.index(t[i])] = '0'
        else:
            answer.append(-1)
    return answer

# 같은 글자를 찾으면 해당 위치를 '0'으로 덮어써서 다음 탐색 시 무시되게 함

# 예) "banana":

# i=3 (a): t[:3] = ['b','a','n']에서 a 찾음 → 거리 3-1=2, 그 자리를 '0'으로 교체 → t = ['b','0','n','a','n','a']
# i=5 (a): t[:5]에서 a 찾음 → 인덱스 3 → 거리 5-3=2
# 단점: t.index()를 두 번 호출하는 비효율이 있음.

# 리스트 index	슬라이싱 + 덮어쓰기	O(n²)
```
