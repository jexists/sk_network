프로그래머스 치킨은 치킨을 시켜먹으면 한 마리당 쿠폰을 한 장 발급합니다. 쿠폰을 열 장 모으면 치킨을 한 마리 서비스로 받을 수 있고, 서비스 치킨에도 쿠폰이 발급됩니다. 시켜먹은 치킨의 수 chicken이 매개변수로 주어질 때 받을 수 있는 최대 서비스 치킨의 수를 return하도록 solution 함수를 완성해주세요.

제한사항

chicken은 정수입니다.

0 ≤ chicken ≤ 1,000,000

입출력 예

| chicken | result |
|--------|--------|
| 100    | 11     |
| 1,081  | 120    |

```python
# 내가 푼 풀이
def solution(chicken):
    answer = -1
    while True:
        if chicken <= 0:
            break
        chicken -= 10
        chicken += 1
        answer += 1

    if(answer == -1):
        answer = 0
    return answer
print(solution(1081))
print(solution(0))
```

### 풀이

```python
def solution(chicken):
    answer = -1
    while True:
        # 치킨이 없는 경우에는 쿠폰 없으니깐 break
        if chicken <= 0:
            break

        # 10개 사면 1개를 줌 (구매한 횟수)
        chicken -= 10 # 10개 구매
        chicken += 1 # 1개 줌
        answer += 1 # 구매한 횟수

    # -1 인 경우 쿠폰 0으로 변경
    if(answer == -1):
        answer = 0
    return answer
print(solution(1081))
print(solution(0))
```
