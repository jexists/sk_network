# 3진법 뒤집기

```python
def solution(n):
    # print(n/3)
    # print(n%3)

    x = []
    while n > 0:
      x.append(n%3)
      n //= 3
    # print(x)
    x.reverse()
    answer = 0
    for i, digit in enumerate(x):
      #  print(f'{digit} * {3**i} = {digit * (3**i)}')
       answer += digit * (3**i)
    return answer



solution(45)
```

### 다른사람 풀이

### 풀이 1 — 문자열 + `int(str, base)` 활용
- `tmp += str(n % 3)` : LSB부터 문자로 붙이면 자동으로 뒤집기 완성 (`reverse()` 불필요)
- `int(tmp, 3)` : 파이썬 내장 기능으로 3진법 문자열 → 10진수 한 번에 변환
- 가장 파이썬다운 풀이로 코드가 짧고 직관적

```python
def solution(n):
    tmp = ''
    while n:
        tmp += str(n % 3)
        n = n // 3

    answer = int(tmp, 3)
    return answer
```

### 내 코드 리뷰

**잘한 점**
- `n > 0` 조건으로 루프 올바르게 종료
- `//=` 정수 나눗셈 사용
- 흐름이 명확하고 읽기 쉬움

**개선 포인트**
- `x.reverse()` 없이도 처리 가능 (Horner's method)
- `3**i` 매 반복마다 누승 계산 → `answer * 3 + d` 방식이 더 효율적

### 풀이 2 — Horner's method (reverse 없이)
- `digits`에 LSB부터 쌓으면 이미 뒤집힌 수의 MSB-first
- `reverse()` 생략 가능, `3**i` 계산 없이 `answer * 3 + d`만으로 변환

```python
def solution(n):
    digits = []
    while n > 0:
        digits.append(n % 3)
        n //= 3
    # digits는 이미 뒤집힌 수의 MSB-first
    answer = 0
    for d in digits:
        answer = answer * 3 + d
    return answer

solution(45)
```

### 풀이 3 — divmod 활용
- `divmod(n, 3)`으로 몫·나머지를 한 번에 처리
- 문자열로 쌓고 `int(..., 3)`으로 마무리

```python
def solution(n):
    digits = []
    while n:
        n, r = divmod(n, 3)
        digits.append(r)
    return int(''.join(map(str, digits)), 3)

solution(45)
```

### 풀이 4 — 한 줄 (코드골프 스타일)
- 세미콜론으로 압축. 가독성은 낮지만 간결함 극대화

```python
def solution(n):
    t = ''
    while n: n, r = divmod(n, 3); t += str(r)
    return int(t, 3)

solution(45)
```
