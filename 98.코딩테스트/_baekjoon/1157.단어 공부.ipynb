### 풀이 과정

````python

a = input()

# 소문자로 변경 (대소문자 구별안함)
list1 = list(a.lower())
# 알파벳을 담는 변수 {'알파벳': 갯수}
dic = {}


for i in list1:
    # 알파벳이 없는 경우 개수 0으로 지정
    if dic.get(i) == None:
        dic[i] = 0
    else:
    # 알파벳이 있는 경우 현재 개수 +1
        dic[i] = dic[i] + 1

# Mississipi

# print(dic) # {'m': 0, 'i': 3, 's': 3, 'p': 0}
count = 0 # 젤 높은 수와 같은 숫자 (가장많이 사용한 숫자가 중복인 경우)
final = '' # 알파벳 (가장 많이 사용한 숫자)

# max_value = max(list(dic.values())) # 리스트 변환 필요없음
max_value = max(dic.values())

# print(dic.values()) # dict_values([0, 3, 3, 0])
# print(max(dic.values())) # 젤 많은 수
# print(max_value) # 3

for i, v in dic.items(): # dic 순환 {알파벳: 숫자}
    if (v == max_value): # 숫자가 젤 높은 수와 같은 경우
        count = count + 1 # 중복이 있는지 여부 확인
        final = i # 알파벳

if count > 1:
    print('?') # 중복이 있으면 '?'
else:
    print(final.upper()) # 대문자로 출력
````

````python
# 3년전 코드.... 
from collections import Counter

# 대문자 출력으로 대문자로 변환
word = input().upper()

# Mississipi

# 단어를 알파벳 순으로 sorted 후
# print(sorted(word)) # ['I', 'I', 'I', 'I', 'M', 'P', 'S', 'S', 'S', 'S']
# counter(): 리스트, 문자열 등의 요소 개수를 딕셔너리 형태로 숫자 세어주는 특수 클래스
counter = Counter(sorted(word))
# print(Counter(sorted(word))) # Counter({'I': 4, 'S': 4, 'M': 1, 'P': 1})
count = dict(counter)
# print(count) # {'I': 4, 'M': 1, 'P': 1, 'S': 4}

# print(count.values()) # dict_values([4, 1, 1, 4])
# print(list(count.values())) # max 함수사용 위해 리스트로 변환 [4, 1, 1, 4]
# print(max(list(count.values()))) # 4
# print(list(count.values()).count(4)) # 4개가 몇개 있는지 2개

if list(count.values()).count(max(list(count.values()))) > 1:
  print("?")
else:
  # max() 함수
  # 딕셔너리나 리스트에서 최빈값(가장 많이 등장하는 값) 또는 최대값을 가진 키
  print(max(count, key=count.get))


dic_1 = {'apple': 3, 'banana': 5, 'cherry': 2}
# 1. 일반적인 max (키 이름 순서상 뒤에 있는 것)
print(max(dic_1)) 
# 결과: 'cherry' (a, b, c 중 c가 가장 크기 때문)
# 2. key=dic_1.get을 사용한 max (숫자가 가장 큰 것의 키)
print(max(dic_1, key=dic_1.get))
# 결과: 'banana' (값이 5로 가장 크기 때문)
````

package main

import (
	"bufio"
	"fmt"
	"os"
	"strings"
)

func main() {
	var word string
	reader := bufio.NewReader(os.Stdin)
	fmt.Fscanln(reader, &word)

	word = strings.ToUpper(word)

	var letters = make(map[uint8]int)
	for i := 0; i < 26; i++ {
		letters[uint8(i)+65] = 0
	}

	for i := 0; i < len(word); i++ {
		letters[word[i]]++
	}

	var maxVal = -1
	var maxKey string
	for key, val := range letters {
		if val > maxVal {
			maxVal = val
			maxKey = string(key)
		} else if val == maxVal {
			maxKey = "?"
		}
	}

	fmt.Println(maxKey)
}
