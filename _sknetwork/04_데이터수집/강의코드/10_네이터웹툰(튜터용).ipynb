```python
# 라이브러리 불러오기
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
# 브라우저 띄우기
url = 'https://comic.naver.com/webtoon'
driver.get(url)
time.sleep(1)
```

```python
# 별점 순으로 이동
css_select = "#container > div.component_wrap.type2 > div.ComponentHead__component_head--O2tPr > div > div > button:nth-child(4)"
driver.find_element(By.CSS_SELECTOR, css_select).click()
```

```python
# 선택자 간결화
# 1. 고유 클래스 사용

css_select = '.ComponentHead__component_head--O2tPr button:nth-child(4)'
driver.find_element(By.CSS_SELECTOR, css_select).click()
```

```python
# 2. 태그 생략 : 태그 이름이 중복되지 않는다면 태그 이름도 생략 가능.

css_select = '.type2 button:nth-child(4)'
driver.find_element(By.CSS_SELECTOR, css_select).click()
```
