```python
# 라이브러리 불러오기
import time 
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
# 페이지 이동
driver.get('https://daum.net')
```

```python
# 브라우저 사이즈 조절
driver.set_window_size(300,400)
```

```python
# 브라우저 사이즈 조절
driver.set_window_size(700,1000)
```

```python
# 브라우저 스크롤 조절
driver.execute_script('window.scrollTo(200, 400)')
```

```python
# alert 다루기
driver.execute_script('alert("파이썬 너무 재밌다!!!")')
```

```python
alert = driver.switch_to.alert
alert.accept()
```

```python
# 키워드 검색
driver.find_element(By.CSS_SELECTOR, '#q').send_keys('파이썬')
```

```python
# 검색 버튼 클릭
driver.find_element(By.CSS_SELECTOR, '#daumSearch > fieldset > div > div > button.btn_ksearch').click()
```

```python
# 더보기 버튼 클릭
selector = '#netizen_more_btn_top'
driver.find_element(By.CSS_SELECTOR, selector).click()
```

```python
# 연관 검색어
selector2 = '#netizen_lists_top > .wsn'
elements = driver.find_elements(By.CSS_SELECTOR,selector2)
len(elements)
```

```python
elements[0].text
```

```python
# 리스트에 20개 담아보기
[el.text for el in elements]
```

```python
# 브라우저 종료
driver.quit()
```
