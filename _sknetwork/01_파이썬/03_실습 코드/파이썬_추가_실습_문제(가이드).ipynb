#1.베스킨라빈스 31 게임 만들기

- 게임의 규칙은 다음과 같습니다
1. 숫자 1부터 시작하며, 마지막으로 31을 말하는 사람이 패배합니다.
2. 각 플레이어는 한 번에 1개에서 3개의 숫자를 말할 수 있습니다.
3. 플레이어가 말한 숫자는 이전 숫자에 이어야 하며, 연속된 숫자를 말해야 합니다.
    - 예: 이전 숫자가 5라면, 6, 6과 7, 또는 6, 7, 8을 말할 수 있습니다.
4. 31을 말한 사람이 패배합니다.

##입력한 숫자만큼 숫자를 출력하는 베라 31게임 진행하는 코드 만들기

<출력 예시 코드>
```
게임 시작!
숫자 몇 개를 부를 건가요? (1~3):2
1!
2!
숫자 몇 개를 부를 건가요? (1~3):4
3!
4!
5!
6!
'
'
'
숫자 몇 개를 부를 건가요? (1~3):3
30!
31!
32!
벌칙 당첨!!
```



````python
# 게임 시작
print('게임 시작!')

# 현재 숫자
current_num = 0

while current_num < 31:
    try:
        num = int(input("숫자 몇 개를 부를 건가요? (1~3): "))

        if (num < 1) or (num > 3):
            print("1에서 3 사이의 숫자를 입력해주세요.")
        else:
            for i in range(num):
                print('{}!'.format(current_num + 1 + i))
            current_num = current_num + num
    except ValueError:
        print("숫자를 입력해주세요.")

if current_num >= 31:
    print("벌칙 당첨!!")
````

#2.가위바위보 게임 만들기

- 내용: 컴퓨터와 사용자가 가위, 바위, 보 게임을 하고 승패를 결정


<출력 예시 코드>
```
가위, 바위, 보 중 하나를 입력하세요: 2
잘못된 입력입니다. '가위, 바위, 보' 중에 다시 입력하세요.
가위, 바위, 보 중 하나를 입력하세요: 가위
컴퓨터의 선택: 가위
비겼습니다!
```



````python
import random

def rock_paper_scissors():
    options = ["가위", "바위", "보"]

    while True:
        user_choice = input("가위, 바위, 보 중 하나를 입력하세요: ")
        if user_choice in options:
            break
        print("잘못된 입력입니다. '가위', '바위', '보' 중에 다시 입력하세요.")

    computer_choice = random.choice(options)
    print(f"컴퓨터의 선택: {computer_choice}")

    if user_choice == computer_choice:
        print("비겼습니다!")
    elif (user_choice == "가위" and computer_choice == "보") or \
         (user_choice == "바위" and computer_choice == "가위") or \
         (user_choice == "보" and computer_choice == "바위"):
        print("이겼습니다! 🎉")
    else:
        print("졌습니다 😢")

rock_paper_scissors()
````

- 확장 아이디어:
    - 점수제 게임(3판 2선승제)
    - 사용자와 컴퓨터의 승률을 기록
