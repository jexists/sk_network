# 4.튜플(Tuple) 자료형

- 튜플은 **리스트와 정말 비슷한 자료형**이다.
- 튜플은 대괄호([ ])가 아닌 **소괄호(( ))** 를 사용해서 표현한다.
- 튜플은 한 번 만들면 **요소의 값을 바꿀 수 없음**이 리스트와 다르다.
- 그 외는 리스트와 같은 특성을 가지므로 이해하기 쉬울 것이다.


## 4.1.튜플 만들기

- 리스트는 대괄호[[ ]]이고, 튜플은 소괄호(( ))인 사실만 기억하면 된다.
- 단, **괄호를 생략할 수 있다**는 점은 리스트와 다르다.
- 리스트에서는 대괄호를 생략할 수 없다.

### 4.1.1.다양한 튜플 만들기

**1) 빈 튜플 만들기**

- 다음은 빈 튜플을 만드는 두 가지 방법이다.

```python
score = ()
score
```

```python
score2 = tuple()
score2
```

**2) 요소가 하나인 튜플 만들기**

- 요소가 하나인 튜플을 만들 때는 조심해야 한다.
- 다음은 요소가 하나인 튜플을 만드는 구문이 아닌 정수형 변수를 선언한 구문이다.

```python
# 튜플을 잘못 만든 예
score1 = (90)

print(score1)
print(type(score1))
```

- 요소가 하나인 튜플은 다음과 같이 **콤마를 추가**해서 만들어야 한다.

```python
score3 = (90,)

print(score3)
print(type(score3))
```

**3) 다양한 형태의 튜플**

```python
score4 = (90,70,60)
score5 = 90, 80, 70
score6 = (90, 70, ('A', 'B'))
```

```python
print(score4)
print(score5)
print(score6)
```

### 4.1.2.리스트 vs. 튜플

- 리스트와 튜플의 차이는 사용하는 괄호가 다르며, 튜플은 요소를 변경할 수 없다.

```python
num = [1,2,3,4,5]

num[0] = 10

num
```

- 다음과 같이 튜플을 만들고 요소 하나를 변경한다.
- 리스트와는 달리 요소를 변경할 수 없다는 오류가 발생한다.

```python
num2 = (1,2,3,4,5)

num2[0] = 10

num2
```

### 4.1.3.튜플 활용

- **괄호를 생략**할 수 있는 튜플의 특징을 이용하면 편리한 부분이 있다.

```python
a = 10
b = 20
c = a, b

c
```

- 우선 여러 변수에 동시에 값을 대입할 수 있다.

```python
a, b, c = 10, 20, 30
```

```python
print('a=', a)
print('b=', b)
print('c=', c)
```

- 두 변수값을 교환할 수 있다.

```python
a, b = 10, 20
print('a=', a, 'b=', b)
a, b = b, a
print('a=', a, 'b=', b)
```

- 함수 리턴값을 두 변수에 대입할 수 있다.

```python
a, b = 45, 20
x, y = divmod(a, b)
print(x, y)
```

## 4.2.튜플 인덱싱과 슬라이싱

- 앞에서 다룬 리스트와 방법과 결과가 같다.

**1) 튜플 인덱싱**

```python
score = (90, 85, 80, 84)

print(score[2])
```

```python
score = (90, 85, 80, 84, ('a', 'b', 'c'))

print(score[4][1])
```

```python
print(score[4][2])
```

**2) 튜플 슬라이싱**

```python
score[1:]
```

## 4.3.튜플 연산

- 앞에서 다룬 리스트와 연산 방법과 결과가 같다.

```python
score1 = (90, 85, 80, 84)
score2 = (9, 8, 80, 8)

score3 = score1 + score2
score3
```

```python
score4 = score1 * 3
score4
```
