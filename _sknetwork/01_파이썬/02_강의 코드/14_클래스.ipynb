## 14.2.클래스의 기본 구조

- `class` 키워드를 사용해서 정의한다.
- 속성(변수)과 메서드(함수)를 정의할 수 있다.

```
class 클래스이름:
    def __init__(self, 속성1, 속성2):
        self.속성1 = 속성1
        self.속성2 = 속성2

    def 메서드이름(self):
        # 동작 정의
        pass

```



**1) 클래스의 주요 규칙**

1. 클래스 이름은 **대문자로 시작**한다.
    - Python에서는 클래스 이름을 항상 대문자로 시작하는 것이 관례이다.

2. 클래스를 만들고 나면 이를 이용해 **객체(실제 물건)를 생성**할 수 있다.
    - 클래스는 설계도라면, 객체는 설계도로 만들어진 실제 물건이다.
    - 예를 들어, car1 = Car() 이렇게 클래스를 호출하면 자동차 객체(car1)가 생성된다.

3. 클래스 안에 **메서드(함수)**가 들어갈 수 있다.
    - 클래스 안의 함수는 객체가 사용할 수 있다.
    - 이때 첫 번째 매개변수로 항상 self를 사용해야 한다.
    - self는 **"지금 이 객체 자신"**을 가리킨다.

4. 객체에서 메서드를 호출할 때 **객체.함수명()**으로 사용한다.
    - 예를 들어, car1.drive()처럼 객체와 함수명을 연결해서 호출하면 된다.

## 14.3.생성자(__init__)
- **생성자**는 객체가 생성될 때 **자동으로 호출되는 특별한 메서드**이다.
    - 이름은 반드시 `__init__`으로 작성해야 한다.
- 생성자는 객체가 만들어질 때 **초기 설정(초기값을 입력하거나 변수 생성)**을 도와준다.

## 14.4.클래스 변수

- **클래스 변수**:  클래스에 속한 변수로, **모든 객체가 공유**한다.
    - 클래스 이름으로 접근 가능하다. (클래스명.변수명)
    - 객체끼리 공유하므로 값이 변경되면 모든 객체에 영향을 미친다.

- **객체 변수** : 객체에 속한 변수로, **각 객체마다 독립적**이다.
    - self를 통해 설정된다. (self.변수명)
    - 객체마다 고유한 값을 가질 수 있다.

- 학생 관리 프로그램을 만들 때, 학생에 대한 정보를 담는 클래스를 만들어 보자
    - ① 학번, ② 이름, ③ 나이, ④ 성별, ⑤ 학과
- 학생 <b>인스턴스</b>를 생성함과 동시에 정보를 초기화할 수 있다.
  * <b>생성자</b>: 인스턴스가 생성될 때 자동으로 실행되는 __init__() 메서드를 의미한다.
* <b>self</b>는 인스턴스(instance) <b>자기 자신</b>을 의미한다.
* 해당 인스턴스가 가지는 값과 함수의 인자 값을 구분하여 이해해야 한다.
* <b>__init__() 함수</b> 내부를 고려해 보자.
  * <b>self.name</b>: 현재 인스턴스의 name 변수
  * <b>name</b>: 함수의 파라미터로 넘어 온 name 변수

* <b>self</b>의 동작 방식에 대하여 바르게 이해할 필요가 있다.
* 우리는 아래와 같이 클래스의 메서드를 호출한다.
* 클래스의 <b>메서드 호출</b>

<pre>
student1.add_age(5)
</pre>

* 이때 student1 → self에 대응되고, 값 5 → offset에 대응된다.
* 클래스 <b>메서드 구현</b>

<pre>
def add_age(self, offset):
    self.age += offset
</pre>
