## 2.3.문자열 인덱싱과 슬라이싱

- 문자열은 문자 요소가 모여서 만들어진 자료형, 즉 **'ABC'는 'A', 'B', 'C' 세 요소가 모인 것**
- <font color="red">인덱스 값은 0부터 시작</font>
- 문자열 인덱싱과 슬라이싱은 이후에 다룰 리스트(List)와 비슷한 특성을 가진다.

```python
str = '대한민국'
print(str)
print(str[0])
print(str[3])
```

**1) 문자열 인덱싱**

- 인덱싱(Indexing): 정수형 위치 인덱스를 사용해 원하는 요소를 조회하는 것
- 첫 번째 문자는 인덱스 0에 해당한다.

```python
s = '동해물과백두산이마르고'

print(s[1])  # 해
print(s[5])  # 두
print(s[-3])  # 마
```

**2) 문자열 슬라이싱**
- 슬라이싱(Slicing) : 범위를 지정해 일정 범위의 요소를 조회하는 것
- 슬라이싱은 보통 두 개의 인덱스로 구성되는데, **변수명[시작 인덱스:끝 인덱스]** 의 형태를 가진다.

```python
s = '동해물과백두산이마르고'

print(s)
print(s[4:6])  # 백두
print(s[:4])   # 동해물과
print(s[-3:])   # 마르고
print(s[:])
```

**[참고] 인덱싱으로 문자열 변경 불가**

- 인덱싱, 슬라이싱을 사용해 문자를 변경할 수는 없다.
- 파이썬에서 문자열은 값을 변경할 수 없기 때문에, 불변(immutable) 객체라고도 한다.

```python
str = '나의살던고향은꽃피는산골'
str[1] = '에'
```

**[참고] 슬라이싱의 범위**
- 문자열 인덱싱을 할 때는 범위를 벗어나는 경우 오류가 발생하지만, 슬라이싱의 경우 부드럽게 처리된다.

```python
str1 = 'I love you'

print(str1[2:100000])
```

## 2.4.문자열 포맷팅(Formatting)

- 문자열 일부분만 바꾸면서 해당 문자열을 재사용할 경우가 있다.
- 예를 들어 문자열에서 이름만 변경하며 사용하는 경우이다.
- 이를 **문자열 포맷팅**이라고 하며, **문자열.format()** 형태로 사용한다.


###2.4.1.% 포맷팅
- % 연산자를 이용하여 문자열에 변수나 값을 포함시킬 수 있다.

- 문자열 포맷 코드
    - %s : 문자열(String)
    - %c : 문자 1개(character)
    - %d : 정수(Integer)
    - %f : 부동소수(floating-point)

```python
# 숫자 바로 대입
print('I eat %d mangos.' %9)
```

```python
# 문자열 바로 대입
print('I eat %s mangos.' %"six")
```

```python
# 2개 이상 값 넣기

print('I ate %d mangos. So I was scik for %s days.' %(20, "six"))
```

```python
num = 20
day = "six"
print('I ate %d mangos. So I was scik for %s days.' %(num, day))
```

###2.4.2.문자열.format()
- 중괄호{}를 사용하여 변수를 표시하고 값을 전달할 수 있다.

**1) 기본 사용법**

- 단순 문자열에서 { }는 아무 의미가 없다.

```python
# 단순 문자열
'이름:{}, 과목:{}, 점수:{}'
```

- 하지만 format() 메소드를 만나면 { }는 큰 의미를 가진다.

```python
# format 사용
'이름:{}, 과목:{}, 점수:{}'.format('홍길동', '수학', 85)     # '이름:홍길동, 과목:수학, 점수:85'
```

```python
# format 사용
print('이름:{0}, 과목:{1}, 점수:{2}'.format('홍길동', '수학', 85))     # '이름:홍길동, 과목:수학, 점수:85'
```

- format() 메소드는 대부분 print() 함수 안에서 사용한다.

**2) 자릿수와 정렬 방식 지정**

- 좀 더 세밀한 출력 형태를 지정하고 싶은 경우 사용한다.

```python
# 소수점 자리수 제한
print('{:.3f}'.format(12.1234567))
```

```python
# 기본 정렬
print('[{0:10}][{1:6}]'.format('홍길동', 1200))
```

```python
# 오른쪽, 왼쪽 정렬
print('[{0:>10}][{1:<6}]'.format('홍길동', 1200))
```

```python
# 가운데 정렬
print('[{0:^10}][{1:^6}]'.format('홍길동', 1200))
```

```python
# 공백을 다른 문자로 채우기
print('[{0:-^10}][{1:_^6}]'.format('홍길동', 1200))
```

## 2.6.문자열 관련 주요 메서드

* 변수(또는 개체) 뒤에 '.함수' 형태로 사용하는 함수를 **메서드(Method)** 라고 부른다.
- 아래 언급된 메서드 외에도 문자열 관련 메서드가 상당히 많으니 관련 자료를 참고하기 바람
- 문자열을 변환하는 메서드를 사용해도 그 결과가 변수에 반영되지는 않는다.

```python
# 대문자로 변환
str = 'I have a Dream'
print(str.upper())
print(str)
```

- 변환 결과를 변수에 반영하려면 메소드가 반환하는 값을 변수에 대입해야 한다.

```python
# 대문자로 변환
print(str.upper())

# 변수에 반영
str = str.upper()
print(str)
```

**1) 대/소문자 변환**

- **capitalize()** 메소드: 문자열 첫 문자만 대문자, 나머지 모두 소문자로 변환
- **title()** 메소드: 단어 첫 문자만 대문자, 나머지는 모두 소문자로 변환
- **upper()** 메소드: 모든 문자를 대문자로 변환
- **lower()** 메소드: 모든 문자를 소문자로 변환

```python
# 대/소문자 변환
print(str.lower())
print(str.title())
print(str.capitalize())
```

**2) 자릿수 지정과 정렬**

- **rjust(x)** 메소드: 지정한 x 만큼의 자릿수를 확보한 후 오른쪽 정렬
- **ljust(x)** 메소드: 지정한 x 만큼의 자릿수를 확보한 후 왼쪽 정렬
- **center(x)** 메소드: 지정한 x 만큼의 자릿수를 확보한 후 가운데 정렬

```python
# 지정한 길이의 문자열이 되도록 공백을 추가하고 정렬
str2 = 'Dream'

print('[' + str2.rjust(10)+']')
print('[' + str2.ljust(10)+']')
print('[' + str2.center(10)+']')
```

**3) 일부 변경 및 제거**

- **replace(x, y)** 메소드: 문자열에 포함된 일부 문자열 x를 문자열 y로 변경
- **strip(x)** 메소드: 문자열에서 양쪽의 공백 또는 특수 문자를 제거

```python
# 문자열 변경('now' --> 'now ')
s = 'nowhere'
s.replace('now', 'now ')
```

```python
# 앞 뒤 빈 문자열, 특수 문자 제거
print('      world'.strip())
print('#world#$$#'.strip())
print('#world#$$#'.strip('$#'))
```

- replace() 메소드를 사용하는 경우가 많으니 꼭 기억합시다.

```python
# 문자열 변수
num = '123,456'

num = int(num.replace(',', ''))
type(num)
```

[문3] 위에서 배운 문자열 관련 메소드를 사용해 다음 주어진 리스트의 요소를 모두 대문자로 바꾸시오.


```python
# 리스트 만들기
happy = ['h', 'a', 'p',  'p', 'y']

# 대문자로 변경
happy = ''.join(happy).upper()
print(happy)
print(type(happy))

# 확인
print(list(happy))
```
