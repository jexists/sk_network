# 8.함수(Function)

- 함수는 입력을 받아 무엇인가를 처리한 후 그 결과를 반환하는 것이다.
- 함수를 사용하는 이유는 전체 소스코드를 기능별로 분리하여 작성하는 것이 효율적이기 때문이다.

- 프로그래밍에서 함수란 반복해 사용할 코드를 미리 정의해 두고 필요할 때 사용하는 모듈이다.
- 함수를 사용하면 같은 로직을 반복해 코딩할 필요가 없어진다.
- 함수는 **def** 문으로 정의한다.
- 함수는 기본적으로 세 가지로 구성된다.

        ① 입력(Input)  
        - 입력 매개변수, 생략가능  
            
        ② 처리
        - 코드들  
            
        ③ 출력(Output)
        - 처리 결과, 생략가능

## 8.1.입력(Input)

```python
# 함수 만들기
def hello():
    print('안녕하세요.')

hello()
```

```python
hello()
```

**1) 입력 매개변수**

- 위 함수는 입력 없이 처리되는 함수이다.


- 이제 이름을 입력 받아 처리하는, 즉 입력 매개변수를 갖는 함수를 만들어보자.

```python
def hello(name):
    print(f'{name}님 안녕하세요. 반갑습니다.')

hello('개똥이')
```

**2) 입력 매개변수 기본값**

- 위 함수는 입력 매개변수가 없으면 오류가 발생한다.
- '매개변수 = 값' 형태로 기본값을 지정할 수 있다.
- 이름을 전달하지 않으면 '여러분'이 기본값으로 사용되게 수정한다.

```python
def hello(name='여러분'):
    print(f'{name}님 안녕하세요. 반갑습니다.')

hello()
hello('홍길동')
```

**3) 입력 매개변수 순서**

- 매개변수 이름으로 지정하면 전달되는 순서를 바꿀 수 있다.
- 매개변수 순서에 맞게 값을 넣으면 매개변수 이름을 생략할 수 있다.

```python
# 함수 만들기
def hello(name, loud = 1):
    if loud == 1:
        print(f"우~~~ 와~~~ {name}님 안녕하세요.")
    else:
        print(f"{name}님 안녕하세요.")
```

```python
# 함수 사용
hello(name='홍길동', loud=1)
hello(loud=0, name='홍길동')
hello('홍길동', 0)              # 매개변수 순서에 맞게
hello('홍길동')                 # 함수 생성 시 loud = 1로 선언했으므로, 안쓰면 기본값인 1이 입력됨
```

**4) 임의의 여러 매개변수**

- 일반적으로 임의의 여러 입력 매개변수가 전달된 경우는 함수 안에서 반복문을 사용해 처리한다.

- `*args`는 함수에 전달된 인수들을 **튜플(tuple)** 형태로 받아서 처리
- `**kwargs`는 함수에 전달된 인수들을 **딕셔너리(dictionary)**로 받아 처리

```python
def hello(*args):
    print(args)

hello('홍길동', '김나래', '이지은')
```

```python
def hello(*args):
    for who in args:
        print(f'{who}님 안녕하세요')

hello('홍길동', '김나래', '이지은')
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}: {value}')

print_info(name='기철이', age=25, city='서울')
```

**5) 임의의 여러 매개변수와 고정 매개변수**

- 임의의 여러 입력 매개변수와 고정된 매개변수를 사용할 수 있다.

```python
def hello(*args, title):
    for who in args:
        print(f'{who}{title} 안녕하세요. 반가워요')

hello('개똥이', '기철이', '기영이', title='선생님')
```

**<연습문제>**

[문1] 숫자를 여러 개 받아(*입력변수) 각각의 숫자를 제곱한 후 모두 더한 값을 출력하는 함수(square_sum)를 만드시오.

```python
nums = (1, 4, 3, 5)

total = 0

for num in nums:
    total += num**2
```

```python
# 함수 만들기
def square_sum(*nums):
    total = 0
    for num in nums:
        total += num**2
    print(total)

# 함수 사용
square_sum(1, 4, 3, 5)
```

- 반환되는 값이 없는 함수의 결괏 값을 받아 출력하면 None이 출력된다.

```python
# 계산기 함수
def calculate(a,b):
    sum_result = a + b
    multi_result = a * b
    return sum_result, multi_result

result1, result2 = calculate(10,5)
print(result1)
print(result2)
```

**<연습문제>**

[문1] 두 숫자를 입력받아 몫과 나머지를 반환하는 divide2 함수를 만드시오.

```python
# 함수 만들기
def divide2(a,b):
    quo_result = a // b
    re_result = a % b
    return quo_result, re_result

# 함수 사용
r1, r2 = divide2(10, 5)
print(r1)
print(r2)
```

```python
# 함수 만들기 2
def divide3(a,b):
    result = [a//b, a%b]
    return result

r1, r2 = divide3(10, 5)
print(r1)
print(r2)
```

## 8.3.lambda 함수
- 이름 없는 함수를 정의할 때 한 줄로 간단하게 생성하여 사용
- def 키워드로 정의된 함수보다 간결
- 복잡한 로직보다는 간단한 연산이나 표현식에 주로 사용
- `lambda 매개변수1, 매개변수2, ... : 표현식`

**1) 기본 예제: 두 숫자의 합**

```python
add = lambda x, y: x+y

result = add(3,5)

result
```

**2) 조건을 이용한 필터링**

```python
# 리스트에서 짝수만 필터링하는 람다 함수
numbers = list(range(1,11))

# filter(함수, 리스트) 리스트 안의 값들을 하나씩 함수에 넣어보고,
# 결과가 True인 값들만 모아서 새로운 리스트로 만들어 줌
evens = list(filter(lambda x: x % 2 ==0, numbers))

evens
```

**3) 정렬 기준으로 사용**

```python
# 리스트 안의 각 문자열의 길이를 기준으로 정렬하는 람다 함수
words = ['apple', 'banana', 'date', 'cherry']

# reverse=True 면 내림차순
sorted_words = sorted(words, key=lambda word: len(word), reverse=True)

print(sorted_words)
```

## 8.4.예외 처리

- 함수 내에서 발생할 수 있는 오류에 대한 사전 처리를 권고한다.
- 다음 함수는 오류에 대한 처리가 없어 함수 사용 시 오류가 발생하는 경우이다.

```python
def divide(a, b):
    result = a / b
    return result

divide(10,0)
```

**1) 입력 오류 처리**

- 나누는 값, 즉 분모가 0인 경우에 대한 처리를 함수에 추가한다.

```python
def divide(a, b):
    if b == 0:
        print('나누는 값이 0이 될 수 없습니다.')
    else:
        result = a / b
        return result

divide(10, 0)
```

**2) 입력 데이터 형식 확인**

- 다음과 같이 데이터 형식을 확인할 수 있다.

```python
a = 123
print(type(a) == int)

b = 'abc'
print(type(b) == str)
```

- 입력되는 데이터 형식이 적절한지 확인하는 구문을 함수에 추가한다.

```python
def divide(a,b):
    if b == 0:
        print('나누는 값이 0이 될 수 없습니다.')
    elif (type(a) != int) or (type(b) != int):
        print('정수 연산만 가능합니다.')
    else:
        result = a / b
        return result

divide(10,0)
divide(10,'ab')
divide(10,5)
```
