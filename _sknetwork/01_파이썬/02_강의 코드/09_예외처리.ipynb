# 9.예외처리(Exception Handling)

- 프로그램을 실행하다 보면 예상하지 못한 오류가 발생할 수 있다.
- 예외처리는 오류가 발생하더라도 프로그램이 비정상 종료되지 않도록 대비하는 방법이다.
- 파이썬에서는 주로 `try`, `except`, `else`, `finally` 문을 사용한다.


## 9.1.예외처리가 필요한 이유

- 사용자가 잘못된 값을 입력할 수 있다.
- 파일이 없거나 네트워크가 끊길 수 있다.
- 0으로 나누는 등 계산 자체가 불가능한 경우도 있다.
- 이런 상황을 미리 고려하면 프로그램이 더 안정적으로 동작한다.


## 9.2.try ~ except 기본 구조

- `try` 블록에는 오류가 발생할 가능성이 있는 코드를 작성한다.
- `except` 블록에는 오류가 발생했을 때 실행할 코드를 작성한다.
- 원칙은 단순하다. **`try` 안에는 정말 위험한 코드만 넣는 편이 좋다.**


- 가능하면 `except:`만 쓰기보다 **어떤 예외인지 명확하게 지정**하는 편이 좋다.
- 그래야 예상하지 못한 다른 오류까지 무조건 숨기지 않게 된다.


## 9.3.여러 종류의 예외 처리
- 하나의 `try` 블록에서 여러 종류의 오류가 발생할 수 있다.
- 모든 에러를 같은 메시지로 처리하면 어디에서 문제가 났는지 알기 어렵다.
- 이때 예외 종류별로 `except`를 나누면 사용자에게 더 정확한 안내를 줄 수 있다.
- 실행 흐름은 다음과 같다.
  - `try` 안 코드 실행
  - 문제 발생 시 알맞은 `except` 블록 실행
  - 문제가 없으면 아래 코드로 계속 진행

## **9.4. 주요 예외 종류**

| 예외 | 설명 | 예시 |
| --- | --- | --- |
| `ValueError` | 값이 잘못됨 | `int('hello')` |
| `TypeError` | 타입이 맞지 않음 | `'hello' + 5` |
| `ZeroDivisionError` | 0으로 나눔 | `10 / 0` |
| `IndexError` | 리스트 인덱스 범위 초과 | `[1, 2, 3][10]` |
| `KeyError` | 딕셔너리에 없는 키 사용 | `{'a': 1}['b']` |
| `FileNotFoundError` | 없는 파일 열기 | `open('없는파일.txt')` |


```python
exception_examples = [
    ('ValueError', ValueError, lambda: int('hello')),
    ('TypeError', TypeError, lambda: 'hello' + 5),
    ('ZeroDivisionError', ZeroDivisionError, lambda: 10 / 0),
    ('IndexError', IndexError, lambda: [1, 2, 3][10]),
    ('KeyError', KeyError, lambda: {'a': 1}['b']),
    ('FileNotFoundError', FileNotFoundError, lambda: open('없는파일.txt', 'r', encoding='utf-8')),
]

for name, error_type, action in exception_examples:
    try:
        action()
    except error_type as e:
        print(f'{name}: {e}')
```

## 9.5.else와 finally

- `else`: 예외가 발생하지 않았을 때 실행된다.
- `finally`: 예외 발생 여부와 상관없이 항상 실행된다.


- `finally`는 파일 닫기, 연결 종료, 정리 작업에 자주 사용된다.


```python
file = None

try:
    file = open('인연.txt', 'r', encoding='utf-8')
    text = file.readline()
    print(text)
except FileNotFoundError:
    print('파일이 존재하지 않습니다.')
finally:
    if file is not None:
        file.close()
        print('파일을 닫았습니다.')
```

## 9.6.raise로 예외 직접 발생시키기

- 상황에 따라 개발자가 직접 예외를 발생시킬 수도 있다.
- 문법적으로는 맞지만 **논리적으로 틀린 값**이 들어오면 직접 예외를 만들 수 있다.
- 예를 들어 나이가 음수이거나 지나치게 큰 값이면 프로그램을 그냥 진행시키면 안 된다.
- 이럴 때 `raise 예외종류('메시지')` 형태를 사용한다.

## 9.8.예외처리 작성 시 주의할 점

- `except:`만 무조건 사용하는 습관은 피하는 것이 좋다.
- 예외를 너무 넓게 잡으면 실제 버그를 놓칠 수 있다.
- 예외를 숨기기만 하지 말고, 원인을 확인할 수 있는 메시지를 남기는 것이 좋다.
- 예외가 발생할 수 있는 범위를 너무 크게 잡지 말고 필요한 부분만 `try`로 감싸는 편이 좋다.
