# 7.제어문(Control Statements)

- **조건 처리**나 **반복 처리**를 통해 코드의 실행 흐름을 제어할 수 있다.
- Python 제어문은 **들여쓰기로 블럭을 지정**하니 이에 주의해야 한다.
- 다양한 로직 작성을 위해 필히 익혀야 하는 내용이다.
- 확장 문법을 알아두면 편리하다.

## 7.1.if 문

- 가장 대표적인 제어문이다.
- 특정 조건에 맞는 경우에만 코드를 실행한다.

**1) 단순 if 문**

- 조건이 참인 경우만 처리한다.
- 조건이 거짓인 경우는 아무 처리도 하지 않는다.

```python
my_score = 73

# 80보다 크거나 같으면 10 더하기
if my_score >= 80:
    my_score += 10

my_score
```

```python
score_list = [75, 80, 70, 95, 90]
score_list2 = []

# 리스트에 요소가 있으면 출력, 없으면 말고
if score_list2:
    print(score_list2)
```

**2) if ~ else 문**

- 조건이 참인 경우와 참이 아닌 경우 각각 처리한다.

```python
my_score = 73

# 80보다 크거나 같으면 Pass, 아니면 Fail
if my_score >= 80:
    my_result = 'Pass'
else:
    my_result = 'Fail'

my_result
```

```python
score_list = [75, 80, 70, 95, 90]

# 리스트에 요소가 있으면 출력, 없으면 '요소 없음' 출력

if score_list:
    print(score_list)
else:
    print("요소 없음")
```

- 처리할 구문이 한 줄이라면 다음과 같이 작성할 수 있다.

```python
if score_list: print(score_list)
else: print("요소 없음")
```

- 처리할 구문이 없거나 이후에 작성할 목적으로 **pass**를 사용할 수 있다.

```python
if score_list:
    pass
else:
    print('요소 없음')
```

**3) if ~ elif ~ else 문**

- 여러 조건 중에서 참인 경우 각각 처리하고, 그렇지 않은 경우 다음 조건으로 넘어가 처리한다.


```python
my_score = 76

# 점수에 따른 학점 주기  (80: B, 70:C, 60:D, F )
if my_score >= 90:
    my_grade = 'A'
elif my_score >= 80:
    my_grade = 'B'
elif my_score >= 70:
    my_grade = 'C'
elif my_score >= 60:
    my_grade = 'D'
else:
    my_grade = 'F'

my_grade
```

## 7.2.while 문

- 조건문이 True인 동안 while문 안의 문장이 반복해서 실행된다.
- 특정 상황에서 반복을 중지하게 while 문 안에서 조건을 제어한다.

**1) 일반 while 문**


```python
count = 0
while count < 3:
    print(count)
    count = count + 1
```

- 문자열 길이를 반환하는 len() 함수가 리스트의 요소 개수를 반환한다.

```python
fruits = ['apple', 'banana', 'cherry']

i = 0
while i < len(fruits):
    print(fruits[i])
    i = i + 1
```

**2) 무한루프 다루기**

- break 문: 반복문 안에서 특정 조건이 되어 반복문을 빠져 나올 때 사용한다.
- continue 문: 반복문을 중단시키지 않고 다음 반복으로 넘어갈 때 사용한다.

```python
total, i = 0, 0

# 100까지 짝수 합 구하기
while True:
    i = i + 1

    # 100보다 크면 종료
    if i > 100:
        break

    # 짝수인지 판별
    if i % 2 == 0:
        total = total + i
    else:
        continue

total
```

- input() 함수를 사용해 사용자의 입력을 받을 수 있다.
- isnumeric() 메소드: 숫자이면 True 아니면 False 반환한다.

```python
# 문자를 입력하면 continue, 0을 입력하면 break, 아니면 10으로 나눈 나머지 표시
while True:
    response = input('숫자를 입력하세요(0 = 종료):')
    if not response.isnumeric():
        continue
    if int(response) == 0:
        break
    result = int(response) % 10
    print("{} 나누기 10의 나머지는 {}입니다.".format(response, result))
```

### 7.3.1.리스트와 for 문

- 반복문과 가장 많이 사용되는 자료형이 리스트형이다.

**1) 기본 반복문**

- 첫 번째 요소부터 마지막 요소까지 코드를 반복 처리한다.

```python
member = ['유재석', '박명수', '정준하', '노홍철', '정형돈']

for men in member:
    print(men)
```

**2) enumerate() 함수 사용**

- 영어 단어 enumerate의 뜻은 '**낱낱이 세다**' 이다.
- 문자열이나 컨테이너 자료형을 입력받아, 순번과 요소를 포함하는 오브젝트로 반환한다.
- 순번과 요소를 반환하여 몇 번째, 어느 요소를 처리하고 있는지 확인 가능하다.

```python
# 순번이 짝수인 경우만 '순번: 이름' 형태로 출력
member = ['유재석', '박명수', '정준하', '노홍철', '정형돈']

for idx, men in enumerate(member):
    if idx % 2 == 0:
        print(idx, men)
```

```python
# 문자열
for idx, val in enumerate("동해물과백두산이"):
    print(idx, val)
```

**3) 반복문 확장(list comprehension)**

- 이 방법을 사용하면 반복문 코드가 매우 간결해진다.
- 처음에는 다소 낯설 수 있으나, 이것이 **Python 스러운 문법**이라 할 수 있다.

```python
even_nums = []

for num in range(1, 11):
    if num % 2 == 0:
        even_nums.append(num)
print(even_nums)
```

```python
even_nums = [num for num in range(1, 11) if num % 2 == 0]

print(even_nums)
```

```python
nums = [0,1,2,3,4]
square = []

# 2를 제곱한 결과를 square 리스트에 추가
for x in nums:
    square.append(x ** 2)
square
```

```python
# 반복문 확장 문법 #1
square = [x ** 2 for x in nums ]
square
```

```python
# 반복문 확장 문법 #2
# 반복문 + 조건문: 짝수인 값의 제곱을 구해서 even_squares 리스트에 넣기

even_squares = [x ** 2 for x in nums if x % 2 ==0]
even_squares
```

### 7.3.2.딕셔너리와 for 문

**1) 기본 반복문**

```python
score = {'홍길동': 85, '한사랑': 95, '일지매': 90}

# key와 value 조회
for key in score:
    value = score[key]
    print(key, value)
```

**2) items() 메소드 사용**

- 딕셔너리를 반복문에 사용할 때 key와 value를 반환해주는 items() 메소드를 많이 사용한다.

```python
for key, value in score.items():
    print(key, value)
```

[문1] 학생 이름과 점수를 갖는 아래 딕셔너리에 대해 반복문을 작성해 다음과 같이 학생들 이름과 점수 합이 출력되게 해보시오.

~~~
홍길동: 350

**3) 반복문 확장**

- 앞에서 '리스트와 for 문'에서 다룬 내용과 동일하게 딕셔너리에 대해 사용할 수 있다.

```python
score = {'홍길동': 85, '한사랑': 95, '일지매': 90}

score
```

```python
# 90점 초과인 학생만 출력
score_over_90 = {}

for key, value in score.items():
    if value > 90:
        score_over_90[key] = value

score_over_90
```

```python
# 90점 초과인 학생만 출력
score_over_90 = {key: value for key, value in score.items() if value > 90}
score_over_90
```

- 다음과 같은 코드로 같은 결과를 얻을 수 있다.

```python
score_over_90 = {key : score[key] for key in score if score[key] > 90}
print(score_over_90)
```
