# QR 코드 생성기
- QR 코드를 생성하고 저장하는 프로그램을 만들기
- 문서에서 문자를 읽어 여러 개의 QR 코드를 자동으로 만들어 보기

```python
# QRcode 패키지 설치
%pip install qrcode
```

```python
import qrcode

# QR 코드에 넣은 데이터
data = '안녕하세요. 반갑습니다.'

# QR 코드 생성
img = qrcode.make(data)

# 이미지 파일로 저장
img.save('QRcode.png')
```

```python
text = 'www.naver.com'

f = open('qrdata.txt', 'w')
f.write(text)
f.close()
```

```python
img = qrcode.make(text)
img.save('QRcode_naver.png')
```

```python
import os

# 데이터가 저장된 파일 경로 - 현재 작업 디렉토리에 qrdata.text 파일이 있다고 가정
file_path = 'qrdata.txt'

# QR 코드 생성 함수
def create_qrcode(data):
    # QR 코드 생성
    img = qrcode.make(data)

    # 이미지 파일로 저장
    img_file_path = f'qrcode_{data}.png'
    img.save(img_file_path)
```

```python
# qrdata.txt 파일에 데이터 읽어오기
# 파일이 없는 경우 에러 메시지를 출력하고 종료

try:
    with open(file_path, 'r') as f:
        for line in f:
            data = line.strip()     # 문자열 양쪽 끝의 공백이나 줄바꿈 문자를 제거
            create_qrcode(data)
except FileNotFoundError:
    print(f"Error:{file_path} 파일을 찾을 수 없습니다. 파일이 있는지 확인하고 다시 시도하세요.")
```
