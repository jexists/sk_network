8) 다음 요구사항에 따라 구문을 순서대로 작성하고 확인하세요.

- tip 데이터프레임 total_bill이 45.0 이상, tip이 6.0 이상인 행을 good 데이터프레임으로 선언하세요.
- good 데이터프레임 인덱스를 초기화하세요.(단, 기존 인덱스는 제거)
- good 데이터프레임을 확인하세요.

```python
# total_bill 45.0 이상, tip 6.0 이상
good = tip[(tip['total_bill'] >= 45.0) & (tip['tip'] >= 6.0)]

# 인덱스 초기화
good.reset_index(drop=True, inplace=True)

# 확인
good.head()
```
