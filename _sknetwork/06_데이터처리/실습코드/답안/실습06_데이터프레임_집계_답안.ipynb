# [실습] 데이터프레임 집계

1) pandas 라이브러리를 pd 별칭을 주어 불러오세요.

```python
import pandas as pd
```

2) 다음 경로의 파일을 읽어 graduate 데이터프레임을 만드세요.
- 파일 : 'Graduate_apply.csv'

```python
path = 'data/Graduate_apply.csv'
graduate = pd.read_csv(path)
```

3) 열의 데이터 형식과 값의 개수 등을 확인하세요.

```python
graduate.info()
```

4) 기초통계정보를 확인하세요.

```python
graduate.describe()
```

5) gre 중앙값을 확인하세요.

```python
graduate['gre'].median()
```

6) rank별 gre 중앙값을 집계하여 tmp 데이터프레임으로 선언하고 확인하세요.

```python
# rank별 gre 중앙값
tmp = graduate.groupby(by='rank', as_index=False)[['gre']].median()

# 확인
tmp
```

7) rank별 gpa 평균을 집계하여 tmp 데이터프레임으로 선언하고 확인하세요.

```python
# rank별 gpa 평균
tmp = graduate.groupby(by='rank', as_index=False)[['gpa']].mean()

# 확인
tmp
```

8) rank, admit별 gpa 평균을 집계하여 tmp 데이터프레임으로 선언하고 확인하세요.

```python
# rank별 gpa 평균
tmp = graduate.groupby(by=['rank', 'admit'], as_index=False)[['gpa']].mean()

# 확인
tmp
```

9) rank별 합격자 수를 집계하여 tmp 데이터프레임으로 선언하고 막대 그래프로 시각화하여 확인하세요.

```python
# rank별 합격자 수
tmp = graduate.groupby('rank', as_index=False)[['admit']].sum()
tmp.head()
```

```python
# 시각화 라이브러리 불러오기
import matplotlib.pyplot as plt

# 시각화
plt.bar(tmp['rank'], tmp['admit'])
plt.show()
```

```python
# [참고]
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'

# 데이터 형식 바꾸기
tmp['rank'] = tmp['rank'].astype(str)

# 재정렬
tmp.sort_values(by='admit', ascending=False, inplace=True)

# 시각화
plt.figure(figsize=(5,3))
plt.bar(x = tmp['rank'], height = tmp['admit'])
plt.title('Admit per Rank', size=20, pad=15)
plt.xlabel('rank')
plt.ylabel('admit')
plt.legend(['admit'])
plt.show()
```

10) rank별 합격률을 집계하여 tmp 데이터프레임으로 선언하고 막대그래프로 시각화해 확인하세요.

- admit 열 값이 0, 1이므로 admit 열 평균이 합격률이 됨

```python
# rank별 합격률
tmp = graduate.groupby('rank', as_index=False)[['admit']].mean()
tmp.head()
```

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt

# 시각화
plt.bar(tmp['rank'], tmp['admit'])
plt.show()
```

```python
tmp['rank'] = tmp['rank'].astype(str)


plt.figure(figsize=(5,3))
plt.bar(x = tmp['rank'], height = tmp['admit'])
plt.title('Admit rate per Rank', size=20, pad=15)
plt.xlabel('rank')
plt.ylabel('admit')
plt.legend(['admit'])
plt.show()
```
