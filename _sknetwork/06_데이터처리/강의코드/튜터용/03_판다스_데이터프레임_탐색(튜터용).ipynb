# 2.데이터프레임 탐색

- 데이터프레임 탐색은 **데이터프레임**의 **키**와 **몸무게**를 재는 것이라 할 수 있다.
- 파일에서 불러온 데이터의 크기, 내용, 분포, 누락된 값 등을 확인할 수 있어야 한다.
- 확인된 내용을 통해 데이터 전처리 필요 여부를 결정한다.
- 데이터를 알아야 데이터를 분석할 수 있다.

```python
import pandas as pd
import numpy as np
```

```python
path = 'data/tips.csv'
tips = pd.read_csv(path)

tips.head()
```

**[tips 데이터 셋 정보]**

- total_bill: 금액
- tip: 팁
- sex: 성별(Male, Female)
- smoker: 흡연 여부(Yes. No)
- day: 요일(Thur, Fri, Sat, Sun)
- time: 시간(Dinner, Lunch)
- size: 고객 수

## 2.3.정렬해서 보기

- 인덱스를 기준으로 정렬하는 방법과 특정 열을 기준으로 정렬하는 방법이 있다.
- **sort_values()** 메소드로 **특정 열**을 기준으로 정렬한다.
- **ascending** 옵션을 설정해 오름차순, 내림차순을 설정할 수 있다.
    - ascending=True: 오름차순 정렬(기본값)
    - ascending=False: 내림차순 정렬

```python
# 단일 열 정렬
tips.sort_values(by='total_bill', ascending=False)
```

```python
# 복합열 정렬
tips.sort_values(by=['total_bill', 'tip'], ascending=[False, True])
```

## 2.4.기본 집계

- 데이터를 좀더 이해하기 위해 고유값, 합, 평균, 최댓값, 최솟값 등을 확인한다.

### 2.4.1.고유값 확인

- 범주형 열(열이 가진 값이 일정한 값인 경우, 성별, 등급 등)인지 확인할 때 사용한다.

**1) 고유값 확인**

- unique() 메소드로 고유값을 확인하며, 결괏값은 배열 형태가 된다.

```python
tips['smoker'].unique()
```

**2) 고유값과 개수 확인**

- value_counts() 메소드로 고유값과 그 개수를 확인하며, 결괏값은 시리즈 형태가 된다.
- dropna 옵션을 생략하거나 dropna=True로 지정하면 NaN 값은 대상에서 제외한다.

```python
# day 열 고유값 확인
tips['day'].unique()
```

```python
# day 열 고유값 개수 확인
tips['day'].value_counts()
```

### 2.4.2.최빈값 확인

- 최빈값은 가장 많이 관측되는 수, 즉 주어진 값 중에서 가장 자주 나오는 값이다.
- 예를 들어, {1, 3, 6, 6, 6, 7, 7, 12, 12, 17}의 최빈값은 6이다.
- mode() 메소드로 최빈값을 확인할 수 있다.
- 최빈값이 여렷인 경우 행으로 구분되어 표시된다.

```python
tips['day'].mode()
```

```python
tips['time'].mode()
```

### 2.4.3.기본 통계 메소드 사용

- 데이터를 1차 집계 한 후 분석을 진행하는 경우가 많으므로 필히 알아두어야 할 내용이다.
- 이후에 배우는 Groupby 기능에서 같이 사용된다.

```python
# 리스트 만들기
stock = [[94500, 92100, 92200, 92300],
         [96500, 93200, 95900, 94300],
         [93400, 91900, 93400, 92100],
         [94200, 92100, 94100, 92400],
         [94500, 92500, 94300, 92600]]
dates = ['2019-02-15', '2019-02-16', '2019-02-17', '2019-02-18', '2019-02-19']
names = ['High', 'Low', 'Open', 'Close']

# 데이터 프레임 만들기
df = pd.DataFrame(stock, index=dates, columns=names)

df.head()
```

image.png

- df.sum()은 df.sum(axis=0) 과 같으며 전체 열 기준으로 합을 집계한다.
- 전체 행 기준으로 집계를 하려면 df.sum(axis=1) 형태로 axis=1을 지정한다.

```python
# 전체 열 기준 합
df.sum(axis=0)
```

```python
# 전체 행 기준 합
df.sum(axis=1)
```

```python
# 전체 열 기준 합
tips.sum(axis=0)
```

```python
# 보통 숫자형 열만 선택해서 사용
tips[['total_bill', 'tip', 'size']].sum()
```

```python
tips.sum(numeric_only=True)
```

```python
# tip 열의 최댓값 조회
tips['tip'].max()
```

```python
# total_bill, tip 열 중앙값 확인
tips[['total_bill', 'tip']].median()
```

```python
# 표준편차
tips[['total_bill', 'tip']].std()
```

```python
# 평균값 확인
tips[['total_bill', 'tip']].mean()
```
