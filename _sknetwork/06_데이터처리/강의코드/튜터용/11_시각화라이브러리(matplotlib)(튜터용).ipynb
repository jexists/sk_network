## 3.차트 꾸미기

- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

### 3.1. 축과 관련된 옵션 사용

**1) x축, y축 이름, 타이틀 붙이기**
- **xlabel()** 함수를 사용하여 그래프의 x축에 대한 레이블 표시
- **ylabel()** 함수를 사용하여 그래프의 x축에 대한 레이블 표시

```python
plt.plot(df['Date'], df['Ozone'])

plt.xticks(rotation=45)                 # x축 값 꾸미기 : 방향을 45도 틀기
plt.xlabel('Date')                      # x축 이름 지정
plt.ylabel('Ozone')                     # y축 이름 지정
plt.title("Daily Airquality")
plt.show()
```

**2) 범례, 그리드 추가**
- 범례(legend)란 데이터의 종류를 표시하기 위한 텍스트를 의미

```python
plt.plot(df['Date'], df['Ozone'], label= 'Ozone')       # lable : 범례 추가를 위한 레이블 값
plt.plot(df['Date'], df['Temp'], label='Temp')

plt.xticks(rotation=45)
plt.legend(loc='upper right')           # 레이블 표시하기, loc=위치
plt.grid()
plt.show()
```

**3) 축 범위 조정하기**
- xlim() : X축이 표시되는 범위 지정 [xmin, xmax]
- ylim() : Y축이 표시되는 범위 지정 [ymin, ymax]
- axis() : X, Y축이 표시되는 범위 지정 [xmin, xmax, ymin, ymax]
- 입력 값이 없으면 데이터에 맞게 자동으로 범위 지정

```python
plt.plot(df['Ozone'])

plt.xlim(0,10)
plt.ylim(0,100)

plt.grid()
plt.show()
```

```python
plt.plot(df['Ozone'])

plt.axis([0, 10, 0, 100])

plt.grid()
plt.show()
```

### **3.2. 라인 스타일 조정하기**


- color=  
    - 'red','green','blue' ...
    - 혹은 'r', 'g', 'b', ...
    - https://matplotlib.org/stable/gallery/color/named_colors.html
- linestyle=
    - 'solid', 'dashed', 'dashdot', 'dotted'
    - 혹은 '-' , '--' , '-.' , ':'
    - https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
- marker=

| marker |	description |
| ---- | ---- |
|"." |	point |
|"," |	pixel |
|"o" |	circle |
|"v" |	triangle_down |
|"^" |	triangle_up |
|"<" |	triangle_left |
|">" |	triangle_right |

- https://matplotlib.org/stable/api/markers_api.html
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html#

```python
plt.plot(df['Date'], df['Ozone'],
         color='b',
         linestyle='dashed',
         marker='1')

plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Ozone')
plt.title('Daily Airquality')

plt.show()
```

## 4.여러 그래프 나눠서 그리기
- 그래프를 표현한 결과물을 한덩이의 그림이라고 하는데 이를 figure라고 한다.
- figure에서는 여러 영역을 나눠 다양한 종류의 그래프를 그릴 수 있다.

![](https://codetorial.net/matplotlib/_images/subplot_03.png)

* row : 고정된 행 수
* column : 고정된 열 수
* index : 순서

**1) 위, 아래로(3행 1열) 그래프 그리기**

```python
df.head()
```

```python
plt.figure(figsize=(8,5))

plt.subplot(3,1,1)
plt.plot('Date', 'Temp', data=df)
plt.ylabel('Temp')
plt.grid()

plt.subplot(3,1,2)
plt.plot('Date', 'Wind', data=df)
plt.ylabel('Wind')

plt.subplot(3,1,3)
plt.plot('Date', 'Ozone', data=df)
plt.ylabel('Ozone')

plt.tight_layout()          # 그래프간 간격을 적절히 맞추기
plt.show()
```

**2) 옆으로 (1행 3열) 그래프 그리기**

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.plot('Date', 'Temp', data=df)
plt.xticks(rotation=40)
plt.ylabel('Temp')
plt.grid()

plt.subplot(1,3,2)
plt.plot('Date', 'Wind', data=df)
plt.xticks(rotation=40)
plt.ylabel('Wind')

plt.subplot(1,3,3)
plt.plot('Date', 'Ozone', data=df)
plt.xticks(rotation=40)
plt.ylabel('Ozone')

plt.tight_layout()          # 그래프간 간격을 적절히 맞추기
plt.show()
```

```python
# for문 이용
plt.figure(figsize=(15,5))
vars = ['Temp', 'Wind', 'Ozone']

for i, v in enumerate(vars):
    plt.subplot(1,3,i+1)
    plt.plot('Date', v, data=df)
    plt.title(v)
    plt.xticks(rotation=40)
    plt.grid()

plt.tight_layout()
plt.show()
    
```

## 5.추가기능

### **5.1. 데이터프레임.plot()**
- 데이터프레임을 쉽게 시각화하는 목적을 가지고 있다.
- 축, 제목, 범례 자동 설정

```python
df.plot(x='Date',
        y=['Temp', 'Ozone'],
        title='Airquality')

plt.grid()
plt.show()
```

**<연습문제>**

아래 조건에 맞게 그래프를 그려 보자.(이번에는 데이터프레임.plot 으로 그려 봅시다.)

### **5.2. 수평선 수직선 추가**

```python
plt.plot(df['Ozone'])

plt.axhline(60, color='r', linestyle='--')
plt.axvline(15, color='b', linestyle='--')

plt.show()
```

### **5.3. 그래프에 텍스트 추가**

```python
plt.plot(df['Ozone'])

plt.axhline(60, color='r', linestyle='--')
plt.axvline(15, color='b', linestyle='--')

plt.text(5, 61, '60')
plt.text(15.5, 20, '15')

plt.show()
```

**<연습문제>**

아래 정보로 그래프를 그려 봅시다.

* 오존 농도를 그래프로 그리고,
* 오존 농도의 평균선(수평선)과 평균값(text)을 추가하시오.

```python
ozone_mean = df['Ozone'].mean()

plt.plot(df['Ozone'])
plt.axhline(ozone_mean, color='r', linestyle='--')
# plt.text(12, 41, ozone_mean)
plt.text(12, 41, f'Average : {ozone_mean:.2f}')
plt.grid(linestyle='--', alpha=0.5)

plt.show()
```

## 6.그래프 저장하기
- **savefig()** 함수를 사용
- 주의할 점은 savefig() 함수를 호출한 이후에 plt.show( )를 호출하면 저장된 이미지가 비어 있을 수 있다.
- 그래서 figure를 저장할 때는 plt.show() 를 사용하지 않는 것이 좋다.

```python
df.plot(x='Date',
        y=['Ozone', 'Temp', 'Wind'],
        title='Airquality',
        xlabel='Date',
        figsize=(12,8))
plt.grid()
plt.legend()

plt.savefig('data/airquality_plot.jpg', format='jpg')
```
