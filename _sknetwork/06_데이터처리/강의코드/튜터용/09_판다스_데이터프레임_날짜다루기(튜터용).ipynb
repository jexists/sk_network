# **8.데이터프레임 날짜 다루기**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **8.1.문자형을 날짜형으로 변경**
- 날짜가 문자형으로 되어있다면 날짜형으로 변경해야 여러가지 날짜 계산을 할 수 있다.
- `pd.to_datetime(컬럼, format='날짜 형식')`
- 시간 지정자 참고 : [링크](https://steadiness-193.tistory.com/171#google_vignette)

|형식|설명|
|:--|:--|
|%Y|0을 채운 4자리 연도|
|%y|0을 채운 2자리 연도|
|%m|0을 채운 월|
|%d|0을 채운 일|
|%H|0을 채운 시간|
|%M|0을 채운 분|
|%S|0을 채운 초|


```python
path = 'data/temperatures.csv'

data = pd.read_csv(path)
```

```python
df = data.copy()

df.head()
```

```python
df.info()
```

```python
pd.to_datetime(df['Date'], format="%Y-%m-%d")
```

```python
df['Date1'] = pd.to_datetime(df['Date'], format="%Y-%m-%d")
df.info()
```

```python
df.head()
```

## **8.2.날짜를 원하는 형식으로 변경**
`데이터컬럼.dt.strftime(날짜형식)`

```python
# 4자리 년-월
df['Date1'].dt.strftime('%Y-%m')
```

```python
# 년-월  시간:분
df['Date1'].dt.strftime('%Y-%m %H:%M')
```

## **8.3.dt 연산자**

|연산자|설명|
|:--|:--|
|year|연도|
|month|월|
|day|일|
|dayofweek|요일(0-월요일, 6-일요일)|
|day_name()|요일을 문자열로|

```python
df['year'] = df['Date1'].dt.year
df['month'] = df['Date1'].dt.month
df['day'] = df['Date1'].dt.day
df['dayofweek'] = df['Date1'].dt.dayofweek
df['dayname'] = df['Date1'].dt.day_name()
```

```python
df.head()
```

## **8.4.날짜 계산**
- day 연산: `pd.Timedelta(day=숫자)`
- month 연산: `DateOffset(months=숫자)`
- year 연산: `DateOffset(years=숫자)`

```python
# 1일 후
df['plus_day1'] = df['Date1'] + pd.Timedelta(days=1)
df.head()
```

```python
# 7일 후
df['plus_day7'] = df['Date1'] + pd.Timedelta(days=7)
df.head()
```

```python
# 1달 후
from pandas.tseries.offsets import DateOffset

df['plus_month1'] = df['Date1'] + DateOffset(months=1)
df.head()
```

```python
# 2년 전
df['minus_year2'] = df['Date1'] - DateOffset(years=2)
df.head()
```

## **8.5.기간 이동 계산**
- 시간에 따라 변화하는 데이터의 추세를 파악하기 위해 사용되는 통계 기법

`컬럼.rolling().집계함수`

```python
df1 = data.copy()
df1.head()
```

```python
# 7일 이동평균
df1['ma7'] = df1['Temp'].rolling(7).mean()
df1.head(10)
```

```python
# 30일 이동 평균
df1['ma30'] = df1['Temp'].rolling(30).mean()
df1.head(30)
```

- 평균 외에 합계, 최솟값, 최댓값 등 다양한 연산이 가능하다.

```python
# 7일 동안 합계
df1['Temp'].rolling(7).sum().head(10)
```

```python
# 7일 동안 최댓값
df1['Temp'].rolling(7).max().head(20)
```

```python
# 7일 동안 최솟값
df1['Temp'].rolling(7).min().head(20)
```
