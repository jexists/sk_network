#1.베스킨라빈스 31 게임 만들기

- 게임의 규칙은 다음과 같습니다
1. 숫자 1부터 시작하며, 마지막으로 31을 말하는 사람이 패배합니다.
2. 각 플레이어는 한 번에 1개에서 3개의 숫자를 말할 수 있습니다.
3. 플레이어가 말한 숫자는 이전 숫자에 이어야 하며, 연속된 숫자를 말해야 합니다.
    - 예: 이전 숫자가 5라면, 6, 6과 7, 또는 6, 7, 8을 말할 수 있습니다.
4. 31을 말한 사람이 패배합니다.

##입력한 숫자만큼 숫자를 출력하는 베라 31게임 진행하는 코드 만들기

<출력 예시 코드>
```
게임 시작!
숫자 몇 개를 부를 건가요? (1~3):2
1!
2!
숫자 몇 개를 부를 건가요? (1~3):4
3!
4!
5!
6!
'
'
'
숫자 몇 개를 부를 건가요? (1~3):3
30!
31!

벌칙 당첨!!
```



````python
print('게임 시작')
num = 1
for i in range(32):
    response = input('숫자 몇 개를 부를건가요? (1~3)')
    if not response.isnumeric():
        continue
    elif int(response) >= 4:
        print('4이상의 숫자는 입력 불가합니다.')
        continue
    elif num >= 30:
        print('벌칙당첨!!')
        break
    else:
        count = int(response)
        for i in range(count):
            print(f'{num}!')
            num += 1
        
````

#2.가위바위보 게임 만들기

- 내용: 컴퓨터와 사용자가 가위, 바위, 보 게임을 하고 승패를 결정


<출력 예시 코드>
```
가위, 바위, 보 중 하나를 입력하세요: 2
잘못된 입력입니다. '가위, 바위, 보' 중에 다시 입력하세요.
가위, 바위, 보 중 하나를 입력하세요: 가위
컴퓨터의 선택: 가위
비겼습니다!
```



````python
import random

game = ['가위', '바위', '보']
number = random.randint(1,3)
computer = game[number]
print(computer)


while True:
    response = input('가위, 바위, 보 중 하나를 입력하세요:')
    if response not in game:
        print('잘못된 입력입니다. \'가위, 바위, 보\' 중에 다시 입력하세요.')
        continue
    elif response == game[number]:
        print('비겼습니다.')
        break
    elif game[number] == '가위' and response == '바위':
        print(f'com {game[number]} 나 {response} 이겼습니다.')
        break
    elif game[number] == '가위' and response == '보':
        print(f'com {game[number]} 나 {response} 졌습니다.')
        break
    elif game[number] == '바위' and response == '보':
        print(f'com {game[number]} 나 {response} 이겼습니다.')
        break
    elif game[number] == '바위' and response == '가위':
        print(f'com {game[number]} 나 {response} 졌습니다.')
        break
    elif game[number] == '보' and response == '가위':
        print(f'com {game[number]} 나 {response} 이겼습니다.')
        break
    elif game[number] == '보' and response == '바위':
        print(f'com {game[number]} 나 {response} 졌습니다.')
        break
    else:
        print(f"com {game[number]} 나 {response} ????")
````

````python
import random

#{'기준': [win:lose]}
game = {'가위':['바위','보'], '바위':['보','가위'], '보':['가위', '주먹']}
# 컴퓨터 랜덤 가위바위보 설정
rand = random.randint(1,3)
computer = list(game.keys())[rand]
# print(game['가위'])
# print(list(game.keys())[1])
# print(computer)
# print(list(game['가위'])[0])

while True:
    response = input('가위, 바위, 보 중 하나를 입력하세요:')
    if response not in game:
        print('잘못된 입력입니다. \'가위, 바위, 보\' 중에 다시 입력하세요.')
        continue
    elif response == computer:
        print(f'비겼습니다. 나: {response} c:{computer}')
        break
    elif response == list(game[computer])[0]:
        print(f'이겼습니다. 나: {response} c:{computer}')
        break
    elif response == list(game[computer])[1]:
        print(f'졌습니다. 나: {response} c:{computer}')
        break
    else:
        print(f"오류!! 나: {response} c:{computer}")
````

- 확장 아이디어:
    - 점수제 게임(3판 2선승제)
    - 사용자와 컴퓨터의 승률을 기록

````python
import random

#{'기준': [win:lose]}
game = {'가위':['바위','보'], '바위':['보','가위'], '보':['가위', '바위']}

# rand = random.randint(1,3)
# computer = list(game.keys())[rand]

user_win = 0
computer_win = 0

# 3판 2선증제 (비길수도 있으니 무한반복)
while True:
    # 컴퓨터 랜덤 가위바위보 설정
    rand = random.randint(0,2)
    print(rand)
    computer = list(game.keys())[rand]
    
    response = input('가위, 바위, 보 중 하나를 입력하세요:')
    if response not in game:
        print('잘못된 입력입니다. \'가위, 바위, 보\' 중에 다시 입력하세요.')
        continue
    elif response == computer:
        print(f'비겼습니다. 나: {response} c:{computer}')
    elif response == list(game[computer])[0]:
        print(f'이겼습니다. 나: {response} c:{computer}')
        user_win += 1
    elif response == list(game[computer])[1]:
        print(f'졌습니다. 나: {response} c:{computer}')
        computer_win += 1
    else:
        print(f"오류!! 나: {response} c:{computer}")
    
    if user_win >= 3:
        print('최종 우승자는 저입니댜!')
        break
    elif computer_win >= 3:
        print('루저 외톨이 상처뿐인 머저리! 최종 우승자는 컴퓨터입니다.')
        break
````
