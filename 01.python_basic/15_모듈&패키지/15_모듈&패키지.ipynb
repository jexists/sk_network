# 모듈(module)
- 파이썬 코드가 들어있는 하나의 `.py` 파일
- 즉, 함수를 만들고 그걸 다른 파일에서 쓸 수 있게 만든 코드 조각이다.


- import 모듈명

```python
import identity_util
```

```python
age = identity_util.user_age(1999)
gender = identity_util.user_gender('991212-1429322')

print(age, gender)
```

```python
import assets.identity_util

age = identity_util.user_age(1999)
gender = identity_util.user_gender('991212-1429322')

print(age, gender)
```

- import 모듈명 as 별칭

```python
import identity_util as id
age = id.user_age(1999)
gender = id.user_gender('991212-1429382')

print(age)
```

- from 모듈명 import 함수

```python
from identity_util import user_age, user_gender

age = user_age(2002)
gender = user_gender('001212-2674525')

print(age)
```

# 패키지(package)
- 패키지는 **모듈을 여러 개 묶어놓은 디렉토리(폴더)**이다.
- 폴더 안에 __init__.py 파일이 있으면 파이썬은 그 폴더를 패키지로 인식함
(Python 3.3부터는 __init__.py 없어도 패키지로 인식됨)

```python
import module.math_util
```

```python

result1 = module.math_util.add(10,5)
result2 = module.math_util.subtract(10,5)

print(f'더하기 {result1}, 빼기 {result2}')
```

- import 패키지명.모듈명 as 별칭

```python
import module.math_util as mm

result1 = mm.add(100, 50)
result2 = mm.subtract(100, 40)

print(f'더하기 {result1}, 빼기 {result2}')
```
