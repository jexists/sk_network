#[참고] input 함수

- input(): 키보드(표준 입력장치)로부터 문자열을 입력 받는다.
- input("문자열"): 안내 메시지(문자열)을 출력하고, 키보드로부터 문자열을 입력 받는다.
- 사용자에게 입력받은 정보를 정수 형태로 처리하기 위하여 int() 함수를 함께 사용할 수 있다.
- 일반적으로 프로그램은 사용자의 입력에 따라서 적절한 결과를 뱉는 방식으로 동작한다.
    - 따라서 입력을 처리하기 위한 input() 함수는 매우 중요하다.

```python
data = input()
print(data)
```

```python
name = input("당신의 이름은? ")
print("입력된 값:", name)
```
