# 8.함수(Function)

- 함수는 입력을 받아 무엇인가를 처리한 후 그 결과를 반환하는 것이다.
- 함수를 사용하는 이유는 전체 소스코드를 기능별로 분리하여 작성하는 것이 효율적이기 때문이다.

- 프로그래밍에서 함수란 반복해 사용할 코드를 미리 정의해 두고 필요할 때 사용하는 모듈이다.
- 함수를 사용하면 같은 로직을 반복해 코딩할 필요가 없어진다.
- 함수는 **def** 문으로 정의한다.
- 함수는 기본적으로 세 가지로 구성된다.

        ① 입력(Input)  
        - 입력 매개변수, 생략가능  
            
        ② 처리
        - 코드들  
            
        ③ 출력(Output)
        - 처리 결과, 생략가능

## 8.5.내장 함수

- 이미 Python이 제공하고 있는 함수를 알고 사용할 수 있어야 한다.

**1) abs(x) 함수**

- x의 절대값을 반환

**2) all(x) 함수**

- x의 요소 모두가 참이면 True를 반환

**3) any(x) 함수**

- x의 요소 중 하나라도 참이면 True를 반환

**4) chr(x) 함수**

- 아스키 코드값 x에 대응하는 문자를 반환

**5) dir(x) 함수**

- 객체 x가 가지고 있는 변수나 메소드 목록을 반환

**6) divmod(x, y) 함수**

- x를 y로 나눈 몫과 나머지를 튜플 형태로 반환

**7) enumerate(x) 함수**

- 리스트, 튜플, 문자열을 받아 인덱스와 값을 반환

**8) eval(x) 함수**

- 실행 가능한 문자열 x를 받아 그 문자열을 실행한 결과를 반환

**9) help() 함수**
- 도움말을 출력. 특정 객체나 함수에 대한 문서를 조회할 수 있음

**10) id() 함수**
- 객체의 고유한 메모리 주소를 반환

**11) input() 함수**

- 사용자로부터 입력을 받는다.(입력 받은 값은 언제나 문자열임)

**12) int(x) 함수**

- x를 정수로 변환

**13) len(x) 함수**

- x 요소들 개수를 반환

**14) list(x) 함수**

- x를 리스트 자료형으로 변환하여 반환

**15) map(f, x) 함수**

- 함수 f에 x 요소를 하나씩 전달해 얻은 결과를 반환

**16) max(x) 함수**

- x 요소들 중 최댓값을 반환

**17) min(x) 함수**

- x 요소들 중 최솟값을 반환

**18) ord(x) 함수**

- 문자 x의 아스키 코드값을 반환

**19) pow(x, y) 함수**

- x를 y번 제곱한 결과를 반환

**20) round(x, y) 함수**
- 실수를 반올림한 결과 반환

**21) range(x, y, z) 함수**

- x 부터 y - 1 까지 z 씩 증가하는 객체를 반환

**22) sorted(x) 함수**

- x 요소를 정렬하여 반환

**23) str(x) 함수**

- x를 문자열로 변환하여 반환

**24) sum(x) 함수**

- x의 요소를 모두 더한 결과를 반환

**25) tuple(x) 함수**

- x를 튜플로 변환하여 반환

**26) zip(x, y) 함수**

- x와 y 요소를 쌍으로 묶어서 반환
