## **10.3. 기본적인 파일 읽고 쓰기**

**1) 파일 생성 및 쓰기**
- w 모드로 파일을 열어 write() 메소드로 파일에 기록한다.
- 이미 같은 이름의 파일이 있다면 제거되고 새 파일이 만들어진다.
- 경로를 지정하지 않으면 현재 작업 디렉토리에 기록된다.
- 새로운 파일이 만들어지며, 같은 이름의 파일이 있으면 제거되고 새로 만들어진다.

```python
# 파일 열기
f  = open('Myfile.txt', 'w')
```

```python
# 파일 쓰기
f.write('안녕하세요?\n')

# 파일 닫기
f.close()
```

```python
name = ['유재석', '박명수', '정준하']

f = open('name.txt', 'w')

# 한 줄에 한명씩 출력
for item in name:
    f.write(item + '\n')

f.close()
```

- with 문
    - 파일 처리나 리소스 관리를 더 안전하고 간결하게 하기 위해 사용된다.
    - `with open(파일명, 모드) as 변수: ` 구문으로 파일을 다룰 수 있다.
    - 이때는 close() 사용할 필요 없다.
    - 파일을 열고 자동으로 닫아주는 역할을 한다. 이를 통해 프로그래머가 실수로 파일을 닫지 않는 문제를 예방할 수 있다.

```python
with open('name2.txt', 'w') as f:
    for item in name:
        f.write(item + '\n')
```

**2) 파일 읽기**
- r 모드로 파일을 열어 read() 메소드로 파일을 읽을 수 있습니다.
- 지정한 파일이 없으면 오류가 발생합니다.

```python
# 파일 열기
f = open('Myfile.txt', 'r')

# 내용 읽기
print(f.read())

# 파일 닫기
f.close()
```

**3) 파일 내용 추가**
- a 모드로 파일을 열어 write() 메소드로 내용을 추가할 수 있다.
- 지정한 파일이 없으면 새 파일이 만들어진다.

```python
# 파일 열기
f = open('Myfile.txt', 'a')

# 내용 추가
f.write('만나서 반갑습니다.')

# 파일 닫기
f.close()
```

- 추가된 내용을 확인한다.

```python
# 파일 열기
f = open('/content/Myfile.txt', 'r')
print(f.read())
f.close()
```

**4) 파일 있으면 오류 발생**
- x 모드로 파일을 열면 같은 파일이 있는 경우 오류를 발생시킨다.
- 파일이 없으면 w 모드로 파일은 연 결과가 된다.
- 파일이 이미 존재하는 경우에 덮어쓰는 것을 방지하고 싶을 때 사용한다.

```python
# 파일 열기(파일이 없을 때 생성)
f = open('Myfile4.txt', 'x')

# 내용 쓰기
f.write('반갑습니다.')

# 파일 닫기
f.close()
```

## **10.4. 파일을 다루는 편리한 메소드**

**1) writelines() 메소드**

- writelines() 메소드를 사용해 행 리스트를 파일에 기록할 수 있다.

```python
# 파일 여러 줄 쓰기
hello = ['안녕하세요\n', '만나서 반가워요\n', '오늘 날씨가 좋아요.\n']

f = open('Myfile5.txt', 'w')
f.writelines(hello)
f.close()
```

```python
# 내용 확인
f = open('Myfile5.txt', 'r')
print(f.read())
f.close()
```

**2) readlines() 메소드**
- readlines() 메소드를 사용해 모든 행을 한 번에 읽을 수 있다.
- 결과는 리스트가 되므로, 반복문으로 한 행 씩 읽을 수 있습니다.

```python
# 파일 한 번에 읽기
f = open('Myfile5.txt', 'r')
result = f.readlines()
f.close()

# 내용 확인
print(result)
```

**3) readline() 메소드**

- readline() 메소드를 사용해 행 단위로 읽을 수 있다.
- 결과는 문자열이 된다.

```python
with open('Myfile5.txt', 'r') as f:
    result = f.readline()           # 파일의 첫 번째 줄을 읽어서 result에 저장

    while result:                   # result에 값이 있으면 반복
        print(result, end='')       # 읽어온 한 줄을 출력
        result = f.readline()       # 다음 줄을 다시 읽어와서 result에 저장
```
