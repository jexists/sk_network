# 6.집합(Set) 자료형

- **집합 연산**(교집합, 합집합, 차집합, 대칭 차집합)을 위한 자료형이다.
- 중복을 허용하지 않아 중복된 원소는 하나만 제외하고 모두 무시된다.
- 원소의 **순서가 의미가 없으므로 인덱싱으로 값을 얻을 수 없다.**

## 6.1.집합 만들기

### 6.1.1.집합 직접 만들기

**1) 정수를 갖는 집합**

**2) 문자열을 갖는 집합**

**3) 여러 자료형을 갖는 집합**

### 6.1.2.집합으로 자료형 변환

- **set()** 함수를 사용해 다른 자료형을 집합으로 변환한다.
- 집합은 요소의 순서가 없으니 변환 후 요소 순서가 어떻게 될 지 알 수 없다.

**1) 문자열 → 집합**

- list(), tuple() 함수로 문자열을 리스트나 튜플로 변환할 수 있다.
- set() 함수로 개별 문자를 요소로 갖는 집합을 만들 수 있다.
- 같은 요소를 갖지 않는 집합의 특징에 따라 **중복된 문자는 제거**된다.

**2) range() → 집합**

- range() 함수로 리스트, 튜플, 집합 등 다양한 자료형을 만들 수 있다.

**3) 리스트 → 집합**

- 리스트가 집합으로 변환되면 중복된 요소는 제거된다.

```python
# 리스트 만들기
member = ['홍길동', '한사랑', '홍길동', '최나영', '이지은']
```

**4) 튜풀 → 집합**

- 역시 중복된 요소는 제거된다.

```python
# 튜플 만들기
member = ('홍길동', '한사랑', '홍길동', '최나영', '이지은')
```

- **set()** 함수를 사용해 리스트나 튜플을 집합으로 변환할 수 있다.
- 위 함수들 기능에 의해 집합 기능을 멋지게 활용할 수 있다.

```python
# 리스트 만들기
nums = [12, 45, 23, 21, 36, 28, 12, 36, 35, 28, 45, 63, 12, 21]
```

## 6.2.집합 연산

- 일반적인 집합 연산 개념이 집합 자료형에 그대로 사용이 가능하다.


### 6.2.1.교집합, 합집합, 차집합

**1) 합집합**

- 합집합은 **|** 또는 **union()** 메소드를 사용해 구한다.

**2) 교집합**

- 교집합은 **&** 또는 **intersection()** 메소드를 사용해 구한다.

**3) 차집합**

- 차집합은 **-** 또는 **difference()** 메소드를 사용해 구한다.

**4) 대칭 차집합**

- 대칭 차집합은 **^** 또는 **symmetric_difference()** 메소드를 사용해 구한다.

## 6.3.집합 관련 메소드

**1) 요소 하나 추가**

- 요소 하나를 추가할 때는 **add()** 메소드를 사용한다.
- 리스트는 add()가 아닌 append() 메소드였음을 기억하자.

**2) 여러 요소 추가**

- 여러 요소를 추가할 때는 **update()** 메소드를 사용한다.
- update() 메소드가 변경이 아닌 추가임을 기억하자.
- 요소를 추가하는 것이 집합을 변경하는 것이라 여기면 될 듯 하다.

**3) 무작위 한 요소 삭제**

- 무작위로 요소 하나를 삭제할 때는 **pop()** 메소드를 사용한다.
- pop() 메소드는 삭제되는 요소를 반환하면서 삭제한다.

**4) 요소 삭제(없으면 오류)**

- 요소를 지정해서 삭제할 때는 **remove()** 메소드를 사용한다.
- 삭제하려는 요소가 없으면 오류가 난다.
- discard() 메소드와 비교하기 바란다.

**5) 요소 삭제(없으면 무시)**

- 요소를 지정해서 하나 삭제할 때는 **discard()** 메소드를 사용한다.
- remove() 메소드와는 달리 삭제하려는 요소가 없어도 오류가 나지 않는다.

**6) 모든 원소 삭제**

- 모든 요소를 삭제할 때는 **clear()** 메소드를 사용한다.
