```python
from google import genai

client = genai.Client(api_key="")

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=input('프롬프트 입력해주세요 : '),
)

print(response.text)
```

```python
from google import genai

client = genai.Client(api_key="")

while True:
    user_input = input("나: ")

    if user_input.lower() in ['quit', 'exit']:
        print('챗봇을 종료합니다.')
        break

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=user_input,
    )

    print("챗봇:", response.text)
    print()

```
