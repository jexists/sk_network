## 2.3.문자열 인덱싱과 슬라이싱

- 문자열은 문자 요소가 모여서 만들어진 자료형, 즉 **'ABC'는 'A', 'B', 'C' 세 요소가 모인 것**
- <font color="red">인덱스 값은 0부터 시작</font>
- 문자열 인덱싱과 슬라이싱은 이후에 다룰 리스트(List)와 비슷한 특성을 가진다.

**1) 문자열 인덱싱**

- 인덱싱(Indexing): 정수형 위치 인덱스를 사용해 원하는 요소를 조회하는 것
- 첫 번째 문자는 인덱스 0에 해당한다.

```python
s = '동해물과백두산이마르고'
```

**2) 문자열 슬라이싱**
- 슬라이싱(Slicing) : 범위를 지정해 일정 범위의 요소를 조회하는 것
- 슬라이싱은 보통 두 개의 인덱스로 구성되는데, **변수명[시작 인덱스:끝 인덱스]** 의 형태를 가진다.

```python
s = '동해물과백두산이마르고'
```

**[참고] 인덱싱으로 문자열 변경 불가**

- 인덱싱, 슬라이싱을 사용해 문자를 변경할 수는 없다.
- 파이썬에서 문자열은 값을 변경할 수 없기 때문에, 불변(immutable) 객체라고도 한다.

**[참고] 슬라이싱의 범위**
- 문자열 인덱싱을 할 때는 범위를 벗어나는 경우 오류가 발생하지만, 슬라이싱의 경우 부드럽게 처리된다.

**1) 기본 사용법**

- 단순 문자열에서 { }는 아무 의미가 없다.

```python
# 단순 문자열
'이름:{}, 과목:{}, 점수:{}'
```

- 하지만 format() 메소드를 만나면 { }는 큰 의미를 가진다.

```python
# format 사용
'이름:{}, 과목:{}, 점수:{}'.format('홍길동', '수학', 85)     # '이름:홍길동, 과목:수학, 점수:85'
```

```python
# format 사용
print('이름:{0}, 과목:{1}, 점수:{2}'.format('홍길동', '수학', 85))     # '이름:홍길동, 과목:수학, 점수:85'
```
