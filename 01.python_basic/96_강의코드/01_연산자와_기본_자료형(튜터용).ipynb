### 3.1.1.변수(Variables)

- **변수**는 값을 담는 그릇 또는 상자라고 할 수 있다.
- 이후 연산 등의 **재사용** 목적을 위해 값을 담아 두는 존재

- 변수 이름은 그 의미를 알 수 있게 적절한 단어를 사용
- 변수에 대입된 값에 의해 변수의 자료형(=데이터 형식)이 결정

### 3.1.3.식별자(Identifier)

- **식별자** : 변수 또는 함수 등에 이름을 붙이기 위해 사용하는 단어
- 식별자는 다음과 같은 **규칙**을 갖는다.
 - 키워드를 사용할 수 없음
 - 언더 바(_)를 제외하고는 특수 문자를 사용할 수 없음
 - 공백을 포함해서는 안됨
 - 숫자로 시작하면 안됨
- 다음은 올바른 식별자의 예
 - age, score, member, member_01, num01, num02
- 다음은 올바르지 않은 식별자의 예
 - if, while, 123, 3Member, My Score, a+b
- 대/소문자를 구분하므로 대/소문자 여부에 신경쓰지 않게 **소문자 사용을 권고**함.
- 또한 언더 바와 영어 단어를 적절히 섞어 가독성이 좋게 하기를 권고함.
 - 안 좋은 예: myname, englishscore
 - 좋은 예: my_name, english_score

## 3.2.숫자형 자료형
- Python은 다양한 형태의 값을 저장하고 처리할 수 있도록 여러가지 자료형을 제공한다.
- 우선 가장 기본이 되는 숫자형 데이터 중 정수형, 실수형에 대해 살펴보자

### 3.2.1.정수형과 실수형


**1) 정수형**

- 음의 정수, 0, 양의 정수(자연수)를 합해 정수형이라 한다.
- **정수 끼리의 연산**결과가 **실수가 될 수 있다.**(특히 나누기 연산, 5 / 2 = 2.5)
- 변수 이름을 보고 자료형을 알 수는 없다.
- **type()** 함수로 어떤 자료형인지 확인해야 한다.

```python
# 정수형 변수
x = 5

# 확인
print(x)
print(type(x))
```

**2) 실수형**

- 소수점을 표시할 수 있는 숫자를 실수형이라고 한다.
- 실수 데이터와 정수 데이터 사이에서 연산을 수행하는 경우, 데이터의 **형 변환(정수 → 실수)**이 일어난다.

```python
# 실수형 변수
y = 2.5

# 자료형 확인
print(type(y))

# 연산 확인
print(y, y+1, y**2)
```

```python
# 정수형 변수
z = 123

# 정수 사이의 연산
num = z/4

# 확인
print(num)
print(type(num))
```

## 3.3.연산자(Operators)

- Python은 수식을 **계산(=산술)**하거나 값을 **비교**하거나 **대입**하는 등의 다양한 **연산자**를 가지고 있다.

### 3.3.1.산술 연산자(Arithmetic Operators)

- **사칙연산(+, -, *, /)**은 일반적으로 알고 있는 연산 방법(더하기 빼기 등)과 같다.
- 값을 더하거나 빼는 등의 처리를 하는 연산자를 **산술 연산자**라고 부름
- **정수**는 소수점이 없는 숫자이며, 음의 정수, 0, 양의 정수를 갖는다. *ex) -10, 0, 15*
- **실수**는 소수점이 있는 숫자이다. *ex) 123.45*

```python
# 정수형 변수
num1 = 26
num2 = 5
```

**1) 사칙 연산자**

```python
# 더하기
num1 + num2
```

```python
# 빼기
num1 - num2
```

```python
# 곱하기
num1 * num2
```

```python
# 나누기
num1 / num2
```

**2) 몫, 나머지, 제곱 연산자**

```python
# 몫
num1 // num2
```

```python
# 나머지
num1 % num2
```

```python
# 제곱
num1 ** num2
```

**3) 연산자 활용**

- **% 연산자**는 짝수 또는 홀수 인지를 판단할 때 사용하는 경우가 많다.
- 조건문(if 문)을 아직 배우지 않았지만 다음 코드의 의미는 파악이 될 것이다.

```python
# 변수 선언
num3 = 20

# 짝/홀수 판단
if num3 % 2 == 0:
    result = '짝수'
else:
    result = '홀수'

# 확인
result
```

- **// 연산자**는 10 단위 또는 100 단위 등의 수를 구할 때 사용하는 경우가 많다.

```python
# 변수 선언
age = 26

# 10으로 나눈 몫을 구한 후 10 곱하기
age = age // 10 * 10

# 확인
age
```

### 3.2.2.print() 함수

- 어떤 처리를 위해 미리 작성되어 모듈화된 코드를 **함수(Function)** 라고 부른다.
- 단순 값, 변수의 값, 연산 결과 등을 확인하기 위해 print() 함수를 사용

```python
# 실수형 변수
c = 10.5
d = 2.5
```

```python
# print() 함수를 사용하지 않은 경우
c % d
c // d
```

```python
# print() 함수를 사용한 경우 #1
print(c % d)
print(c // d)
```

- print() 함수 안에 콤마로 구분해 값을 여러 개 전달할 수 있어 다음과 같이 사용할 수 있다.

```python
# print() 함수를 사용한 경우 #2
print('c =', c)
print('d =', d)
print('c % d =', c % d)
```

- print() 함수에 sep, end 옵션을 사용할 수도 있다.

```python
# 구분자 지정
print('Hello', 'world')
print('Hello', 'world', sep=',')
print('Hello', 'world', sep='')
```

```python
# 개행 여부 지정
print('Hello', 'world', end='\n')
print('Hello', 'world')

print('------------------')

print('Hello', 'world', end=' ')
print('Hello', 'world')
```

### 3.2.3.복합 연산자(Assignment Operators)

- **연산과 대입을 동시**에 수행할 수 있는 연산자를 복합 연산자라고 부른다.
- 구문은 짧으나 가독성이 떨어지므로 너무 많이 사용하면 오히려 복잡해 질 수 있다.

```python
# 정수형 변수 선언
num4 = 10

# 복합 대입 연산자 사용
num4 += 10
num4 -= 2
num4 *= 3
num4 /= 5

# 확인
num4
```

- 위 구문은 다음 구문과 같다.

```python
# 정수형 변수 선언
num4 = 10

# 복합 대입 연산자 사용 안함
num4 = num4 + 10
num4 = num4 - 2
num4 = num4 * 3
num4 = num4 / 5

# 확인
num4
```

## 3.4.[참고] 날짜 시간 자료형

- Python은 날짜와 시간 관련 자료형을 제공하지 않는다.
- 필요하다면 **datetime** 라이브러리를 불러와 사용해야 한다.

```python
# 라이브러리 불러오기
from datetime import  datetime

# 현재 날짜와 시간 얻기
dt = datetime.now()
dt
```

```python
# 날짜 시간 정보 표시
print('날짜와 시간: ', dt)
print('년도:', dt.year)
print('월:', dt.month)
print('일:', dt.day)
print('시:', dt.hour)
```
