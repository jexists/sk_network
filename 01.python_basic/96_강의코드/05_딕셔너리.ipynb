# 5.딕셔너리(Dictionary) 자료형

- 딕셔너리는 **중괄호({ })** 를 사용해 선언한다.
- {Key1: Value1, Key2: Value2, Key3: Value3}의 형태를 가진다.
- 딕셔너리는 순서의 의미가 없으며, **키(Key)를 사용해 값(Value)을 확인한다.**
- 리스트와 더불어 많이 사용되는 자료형이다.

## 5.1.딕셔너리 만들기

### 5.1.1.딕셔너리 직접  만들기

**1) 일정한 정보를 갖는 딕셔너리**

**2) 다양한 정보를 갖는 딕셔너리**

- 리스트나 튜플을 Value로 가질 수 있다.

**3) 숫자가 Key인 경우**

- 당연히 숫자도 Key가 될 수 있다.

**4) 같은 Key 무시**

- 만일 Key가 같은 요소가 있으면 하나만 남기도 나머지는 무시된다.

**[참고] Key가 될 수 없는 리스트**

- 당연히 Key는 변경되면 안된다.
- 튜플은 변경 불가능한 자료형이지만, **리스트는 변경 가능**한 자료형이다.
- 그러므로 리스트는 딕셔너리의 Key가 될 수 없다.

**[참고] 또 다른 방법**

- dict 함수를 사용해 다음과 같이 만들 수도 있다.

### 5.1.2.딕셔너리로 자료형 변환

- **dict() 함수**를 사용해 Key:Value 형태를 가질 수 있는 자료형을 딕셔너리로 변환한다.
- 대부분 리스트로 딕셔너리를 만든다.

**1) 리스트 → 딕셔너리 변환**

- [Key, Value] 형태의 요소를 갖는 리스트를 딕셔너리로 변환해 보자.

````python
members = [['홍길동', 100], ['김철수', 90], ['최나영', 90], ['이지수', 95]]
````

**2) 튜플 → 딕셔너리 변환**

- 당연히 튜플을 딕셔너리로 변환할 수 있다.

````python
members = (('홍길동', 100), ('김철수', 90), ('최나영', 90), ('이지수', 95))
````

## 5.3.딕셔너리 변경, 추가와 삭제

- 인덱스 대신 Key로 찾아 변경한다는 것 외에는 리스트와 같다.

**1) 변경, 추가**

````python
member = {'이름': '홍길동',
          '나이': 20}
````


- 없는 Key로 변경하면 Key:Value 형태로 추가된다.

**2) 삭제**

````python
member = {'이름': '홍길동',
          '나이': 20,
          '지역':'서울',
          '성별': 'M',
          '등급': 'Gold'}
````

- **del** 문으로 Key를 지정해 해당 요소를 삭제한다.

- **pop()** 메소드를 사용하면 Key를 지정해 요소를 삭제할 수 있다.
- 그리고 삭제된 요소의 Value를 반환받는다.

- **popitem()** 딕셔너리에서 마지막에 추가된 (key, value) 쌍을 꺼내면서 동시에 삭제하는 메서드이다.

## 5.4.딕셔너리 관련 메소드

**1) Key 확인**

- **keys()** 메소드를 사용해 Key 정보를 확인한다.
- 모든 Key를 요소로 갖는 dict_keys 개체를 반환한다.

````python
member = {'이름': '홍길동',
          '취미': ['독서', '여행', '걷기']}
````

**2) Value 확인**

- **values()** 메소드를 사용해 Value 정보를 확인한다.
- 모든 Value를 요소로 갖는 dict_values 개체를 반환한다.

**3) Key, Value 확인**
- **items()** 메소드를 사용해 Key와 Value 정보를 확인한다.
- 모든 key와 value를 (key, value) 형태로 갖는 dict_items 개체를 반환한다.

**4) 모든 요소 삭제**
- **clear()** 메소드를 사용해 모든 요소를 삭제한다.

**5) 찾기(없을 때 대신할 값 지정)**

- **get()** 메소드를 사용해 Key를 지정해 Value를 찾을 수 있다.
- Key로 Value를 찾을 수 없을 때 대신 표시할 값을 지정할 수 있다.

````python
member = {'이름': '홍길동', '취미': ['독서', '여행', '걷기']}
````

- 대응되는 Value를 찾지 못할 경우 반환되는 값을 '없음'으로 지정할 수 있다.
