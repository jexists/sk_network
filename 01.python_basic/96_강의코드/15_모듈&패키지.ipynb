# 모듈(module)
- 파이썬 코드가 들어있는 하나의 `.py` 파일
- 즉, 함수를 만들고 그걸 다른 파일에서 쓸 수 있게 만든 코드 조각이다.


- import 모듈명

- import 모듈명 as 별칭

- from 모듈명 import 함수

# 패키지(package)
- 패키지는 **모듈을 여러 개 묶어놓은 디렉토리(폴더)**이다.
- 폴더 안에 __init__.py 파일이 있으면 파이썬은 그 폴더를 패키지로 인식함
(Python 3.3부터는 __init__.py 없어도 패키지로 인식됨)

- import 패키지명.모듈명 as 별칭
