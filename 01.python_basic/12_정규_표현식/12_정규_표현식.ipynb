## 11.1.정규식 사용 전/후 비교

- 기존의 반복문으로 처리하던 것을 정규 표현식으로 처리하면 코드가 단순해진다.
- 물론 정규식에 대해 알지 못하면 암호(?)같은 느낌일 것이다.
- 우선 다음과 같이 이름과 전화번호가 섞여 있는 문자열이 있다.

```python
# 이름과 전화번호가 섞여 있는 문자열
peoples = '영수: 010-1234-5678 영철: 010-4321-8765 옥순: 010-2143-6587'
```

- 위 문자열에서 전화번호 뒤 네 자리를 XXXX로 변경하려고 한다.
- 기존 반복문과 정규식을 사용하는 방법 중 무엇이 간결한지 비교한다.

**1) 기존 반복문으로 처리**

- 기존 반복문으로 전화번호 뒤 네 자리를 XXXX로 변경한다.

```python
# 전화번호 뒷자리 숨기기
phone_x = []
for word in peoples.split(' '):
    if len(word) == 13 and word[:3].isdigit() and word[4:8].isdigit() and word[9:13].isdigit():
        word = word[0:8] + '-XXXX'
        phone_x.append(word)
    else:
        phone_x.append(word)
peoples_new = ' '.join(phone_x)
print(peoples_new)
```

**2) 정규식으로 처리**

- 정규식으로 전화번호 뒤 네 자리를 XXXX로 변경한다.



```python
# 전화번호 뒷자리 숨기기

import re

# re.sub(패턴, 바꿀값, 대상문자열)
re.sub('(\d{3}[-]\d{4})[-]\d{4}','\g<1>-xxxx', peoples)
```

- 위 구문은 다음과 같이 작성할 수도 있다.
- 특정 패턴을 이후에 반복해서 사용할 때 좋은 방법이다.

```python
# 전화번호 뒷자리 숨기기
pat = re.compile('(\d{3}[-]\d{4})[-]\d{4}')
pat.sub('\g<1>-XXXX', peoples)
```

## 11.2.기본적인 정규식

- 정규식을 사용하려면 re 라이브러리를 불러와야 한다.

```python
# 모듈 불러오기
import re
```

- 다음과 같은 메소드를 사용해 특정 패턴의 문자열을 확인할 수 있다.
    - match() 메소드: 문자열 처음부터 정규식과 매치 여부를 확인한다.
    - search() 메소드: 문자열 전체를 대상으로 정규식과 매치 여부 확인한다.
    - findall() 메소드: 정규식과 매치되는 모든 문자열을 리스트 형태로 반환한다.
    - finditer() 메소드: 정규식과 매치되는 모든 문자열을 반복 가능한 개체로 반환한다.

match vs search

re.match(): 문자열의 첫 번째 글자부터 패턴이 일치하는지 확인합니다. 시작이 다르면 바로 None을 줍니다.

re.search(): 문자열 전체를 훑어서 패턴이 어디에든 있는지 확인합니다

### 11.2.1.특정 문자 또는 문자열 포함 여부 확인

- match() 메소드는 문자열 처음부터 정규식와 매치 여부를 확인한다.

```python
# 맨 앞에 'I' 문자가 있는지 확인
string = 'i have a Dream'

print(re.match('i', string))
print(re.match('I', string))

print(re.match('i hav', string))

print(re.match('ha', string))
```

- search() 메소드는 문자열 전체를 대상으로 정규식과 매치 여부를 확인한다.

```python
# 문자열에 'a' 문자가 있는지 확인

string = 'I have a Dream'
print(re.search('a', string))
print(re.search('hav', string))
```

- 조건문을 사용해 문자열에서 특정 문자 또는 문자열이 존재하는 지 표시할 수 있다.

```python
string = 'I have a Dream.'
result = re.search('a',string)

if result:
    print('찾음')
else:
    print('못찾음')
```

- 매치된 결과를 다음과 같은 메소드로 확인할 수 있다.
    - group():	매치된 문자열을 반환한다.
    - start():	매치된 문자열의 시작 위치를 반환한다.
    - end():	매치된 문자열의 끝 위치를 반환한다.
    - span():	매치된 문자열의 (시작, 끝)에 해당하는 튜플을 반환한다.

```python
# 'ha' 문자 찾기
string = 'I have a Dream.'
result = re.search('ha', string)

# 매칭된 문자 출력
print(result.group())

# 매칭된 문자 시작 위치
print(result.start())

# 매칭된 문자 끝 위치, 매칭된 부분의 끝 위치 +1 (인덱스)
print(result.end())

# 매칭된 구간 (start, end) 튜플
print(result.span())
```

```python
# 'Dream' 문자열 찾기

string = 'i hava a Dream'
result = re.search('Dream', string)

print(result.group())
print(result.start())
print(result.end())
print(result.span())
```

- findall() 메소드는 정규식과 매치되는 모든 문자열을 리스트 형태로 반환한다.

```python
# 'a' 문자 찾기
string = 'i have a Dream'
re.findall('a', string)
```

```python
# 'ave' 문자열 찾기

re.findall('ave',string)
```

- finditer() 메소드는 정규식과 매치되는 모든 문자열을 반복 가능한 객체로 반환한다.
- 반복 가능하다는 것은 반복문에 사용가능하다는 의미이다.

```python
# 반복문으로 표시

string = 'i have a Dream'
result = re.finditer('a', string)

for i in result:
    print(i)
```

### 11.2.2.나열된 문자 포함여부 확인

- 대괄호([ ]) 안에 문자를 나열하여 그 문자들 포함 여부를 확인한다.

```python
# 'a', 'h', 'm' 문자 찾기
string = 'I have a Dream'
re.findall('[ahm]',string)
```

```python
# 지우고 싶은거 나열
re.sub('[ahm]','',string)

re.sub('[ahm]','1',string)
```

```python
# 'a', 'v', 'e' 문자 찾기

string = 'I have a Dream'
re.findall('[ave]', string)
```

```python
# 같은 문자 한 번만 표시
result = re.findall('[ave]', string)

result = set(result)

print(list(result))

result = list(set(re.findall('[ave]', string)))
```

- ^를 사용해 나열되지 않은 문자들 포함 여부를 확인할 수 있다.

```python
# 'a', 'v', 'e' 이외의 문자 찾기
import re
string = 'i have a Dream'
result = list(set(re.findall('[^ave]', string)))
result
```

### 11.2.3.범위 안의 문자 포함여부 확인

- 빼기 기호(-)를 사용해 문자의 범위를 지정한다.

```python
# 'a' ~ 'm' 문자 찾기
re.findall('am', string)

re.findall('[a-m]',string)

```

```python
# 'a' ~ 'm' 문자 찾기(대/소문자 무시)
re.findall('[a-mA-M]',string)
```

```python
# 'a' ~ 'm' 이외의 문자 찾기(대/소문자 무시)
re.findall('[^a-mA-M]', string)
```

```python
# 숫자 찾기
string2 = 'i am 10 years old'

print(re.findall('\d', string2))

print(re.findall('[0-9]',string2))
```

### 11.2.4.두 문자 사이의 문자 포함여부 확인

- 정규식의 Dot(.) 메타 문자는 줄바꿈 문자인 \n을 제외한 모든 문자와 매치된다.

```python
# 'a'와 'e'문자 사이에 문자 하나를 포함하는 문자열 확인
string = 'I have a Dream'
print(re.findall('a.e', string))

print(re.findall('a...e', string))
```

```python
# 'r'과 'm'문자 사이에 문자 하나를 포함하는 문자열 확인
print(re.findall('r.m', string))
print(re.findall('r..m', string))
```

### 11.2.5.문자 반복 확인

- 문자의 반복을 확인하는 여러 메타 문자가 있다.

```python
# o가 0회 이상 반복된 경우
string = 'cl col Coor coool pool cooool'

re.findall('co*l', string)
```

```python
# o가 1회 이상 반복된 경우

re.findall('co+l', string)
```

```python
# o가 1회 반복된 경우

re.findall('co+l', string)
```

```python
# o가 1 ~ 4회 반복된 경우
# o가 n회 반복 여부 확인
print(re.findall('co{3}l', string))

print(re.findall('co..l', string))
```

```python
# o가 1-4회 반복된 경우
re.findall('co{1,4}l',string)
```

```python
# o가 0-1회 반복된 경우
re.findall('co?l',string)
```
