```python
import os

# 데이터가 저장된 파일 경로 - 현재 작업 디렉토리에 qrdata.text 파일이 있다고 가정
file_path = 'qrdata.txt'

# QR 코드 생성 함수
def create_qrcode(data):
    # QR 코드 생성
    img = qrcode.make(data)

    # 이미지 파일로 저장
    img_file_path = f'qrcode_{data}.png'
    img.save(img_file_path)
```

```python
# qrdata.txt 파일에 데이터 읽어오기
# 파일이 없는 경우 에러 메시지를 출력하고 종료

try: 
    with open(file_path,'r') as f:
        for line in f:
            data = line.strip()
            create_qrcode(data)
except FileNotFoundError:
    print(f'Err:{file_path} 파일을 찾을 수 없습니다. 파일이 있는지 확인')
```

2. 공식문서를 보고 box_size 등 여러 옵션을 달리해서 qr

```python
import qrcode
my_info = {'이름': '정주애', '전화번호': '010-1111-2222', '이메일': 'jungia21@naver.com', 'portfolio':'https://github.com/jexists'}


text = ''
for i, v in my_info.items():
    text += f'{i}: {v} \n'

qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=1,
)
qr.add_data(text)
qr.make(fit=True)

# img = qr.make_image(fill_color="black", back_color="white")
img = qr.make_image(back_color=(0, 0, 0), fill_color=(255, 255, 255))
img_file_path = f'my_info.png'
img.save(img_file_path)
```
