# 3.리스트(List) 자료형

- 리스트는 여러 값을 묶어서 갖는 컨테이너 자료형 중의 하나이다.
- **대괄호([ ])** 안에 콤마로 구분해 값을 나열하며, 이 값을 **요소**라고 부른다.
- 이 세상 온갖 데이터(숫자, 문자, 파일, 이미지 등등)가 리스트의 요소가 될 수 있다.
- 당연히 리스트가 리스트의 요소가 될 수도 있다.
- Python에서 가장 많이 사용되는 자료형이 리스트라고 생각해도 된다.

**[참고] Python과 괄호**

- Python은 여러 값을 묶어서 가질 수 있는 **컨테이너 자료형**을 제공한다.
- **리스트(List), 튜플(Tuple), 집합(Set), 딕셔너리(Dictionary)**가 컨테이너 자료형이다.
- 여러 값을 묶기 위해서는 당연히 **괄호**가 필요하다.
- 어떤 괄호를 사용하는가에 따라 자료형이 결정된다.
- Python은 우리가 아는 모든 괄호를 다 사용한다.
    - **대괄호**로 값을 묶으면 **리스트**가 된다. ex) [1, 2, 3, 4]
    - **소괄호**로 묶으면 **튜플**이 된다. ex) (1, 2, 3, 4)
    - **중괄호**로 묶으면 **집합**이 된다. ex) {1, 2, 3, 4}
    - 그런데 **딕셔너리**에 사용할 더 이상의 괄호가 없다. 그래서...

- 이러한 인덱스로 특정 요소를 찾는 과정을 **인덱싱**이라고 한다.

- 인덱스 역방향은 **-1**부터 시작된다.
