```python

# 타입이 아니라 "값"을 비교하기 때문
# int와 float를 자동으로 맞춰서 비교
print(int('1') == 1)
print(float('1') == 1)

print(0.1 == 1)
```

```python
print(type(float('1')))  # float
print(type(1))           # int

print(float('1') == 1)   # True
print(type(float('1')) == type(1))  # False
```
