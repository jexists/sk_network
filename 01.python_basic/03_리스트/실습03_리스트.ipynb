5) 리스트 a에 23, 3, 16, 45, 11을 저장하고, 아래 예시와 같이 출력하세요.

- 첫 번째 줄은 순방향 인덱싱과 슬라이싱을 사용하세요.
- 두 번째 줄은 역방향 인덱싱과 슬라이싱을 사용하세요.
~~~
[3, 16, 45] [23, 3, 16] [16, 45] [16, 45, 11]
[16, 45] [] [16, 45, 11] [23]
~~~

```python
a = [23,3,16,45,11]

print(a[1:4], a[:3], a[2:4],a[2:])


print(a[-3:-1], a[-1:-1], a[-3:],a[-5:1])
```

# 실습 #3: 리스트 수정하기
