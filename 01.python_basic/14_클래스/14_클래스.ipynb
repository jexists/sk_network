# 14.클래스(Class)

- 특정 데이터와 해당 데이터에 대한 행동(메서드)을 정의하는 **설계도** 같은 역할을 한다.
- 클래스는 프로그램 작성을 위해 꼭 필요한 요소는 아니다. 하지만 프로그램을 작성할 때 클래스를 잘 사용한다면 프로그래머가 얻을 수 있는 이익은 많다.


## 14.1.클래스를 사용하는 이유

- **재사용성**: 반복적으로 사용하는 데이터를 하나로 묶어서 관리할 수 있다.
- **캡슐화**: 내부 데이터를 외부에서 직접 변경하지 못하게 보호할 수 있다.
- **추상화**: 복잡한 동작을 단순하게 표현 가능하다.
- **상속과 다형성**: 기존 클래스의 기능을 확장하거나 변경해 새로운 동작을 추가할 수 있다.

````python
# 계산기의 '더하기' 기능 구현한 파이썬 코드

result = 0

def add(num):
    global result
    result += num
    return result

print(add(5))

print(add(10))
````

- 만약 2대의 계산기가 필요하다면

````python

result1 = 0
result2 = 0

# 계산기1
def add1(num):
    global result1
    result1 += num
    return result1

# 계산기2
def add2(num):
    global result2
    result2 += num
    return result2

print(add1(5))
print(add1(10))


print(add2(50))
print(add2(100))

````

- 클래스를 사용해서 구현

````python
# 클래스 정의하기

class Calculator:
    # 초기 설정 함수
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

# 객체 만들기
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(4))
print(cal1.add(3))

print(Calculator().add(3))
print(Calculator().add(3))

print(cal2.add(4))
print(cal2.add(4))
````

    - 예를 들어, car1.drive()처럼 객체와 함수명을 연결해서 호출하면 된다.

## 14.4.클래스 변수

- **클래스 변수**:  클래스에 속한 변수로, **모든 객체가 공유**한다.
    - 클래스 이름으로 접근 가능하다. (클래스명.변수명)
    - 객체끼리 공유하므로 값이 변경되면 모든 객체에 영향을 미친다.

- **객체 변수** : 객체에 속한 변수로, **각 객체마다 독립적**이다.
    - self를 통해 설정된다. (self.변수명)
    - 객체마다 고유한 값을 가질 수 있다.

````python
class School:
    # 클래스 변수
    school_name = '파이썬 고등학교'
    
    # 생성자: 객체 변수를 초기화
    def __init__(self, student_name, strudent_age):
        self.student_name = student_name
        self.strudent_age = strudent_age

# 객체 생성
student1 = School('최영철', 15)
student2 = School('김옥순', 16)

# 객체 변수 출력
# print(School.student_name, School.strudent_age)
print(student1.student_name, student1.strudent_age)
print(student2.student_name, student2.strudent_age)

# 클래스 변수 출력
print(School.school_name)
print(student1.school_name)
print(student2.school_name)
````

- 학생 관리 프로그램을 만들 때, 학생에 대한 정보를 담는 클래스를 만들어 보자
    - ① 학번, ② 이름, ③ 나이, ④ 성별, ⑤ 학과
- 학생 <b>인스턴스</b>를 생성함과 동시에 정보를 초기화할 수 있다.
  * <b>생성자</b>: 인스턴스가 생성될 때 자동으로 실행되는 __init__() 메서드를 의미한다.
* <b>self</b>는 인스턴스(instance) <b>자기 자신</b>을 의미한다.
* 해당 인스턴스가 가지는 값과 함수의 인자 값을 구분하여 이해해야 한다.
* <b>__init__() 함수</b> 내부를 고려해 보자.
  * <b>self.name</b>: 현재 인스턴스의 name 변수
  * <b>name</b>: 함수의 파라미터로 넘어 온 name 변수

````python
# 학생 관리 프로그램
# 학생 id, 이름, 나이, 성별, 전공

class Student:
    def __init__(self, id, name, age, gender, department):
        self.id = id
        self.name = name
        self.age = age
        self.gender = gender
        self.department = department

    def show(self):
        print('=== 학생정보 ===')
        print(f'학번:{self.id}')
        print(f'이름:{self.name}')
        print(f'나이:{self.age}')
        print(f'성별:{self.gender}')
        print(f'학과:{self.department}')

    def add_age(self, offset):
        self.age += offset
        

student1 = Student('20200001','최영철', 20, '남성','컴퓨터공학과')
student2 = Student('20200002','김옥순', 22, '여성','시각디자인학과')

student1.show()
student2.show()
````

* <b>self</b>의 동작 방식에 대하여 바르게 이해할 필요가 있다.
* 우리는 아래와 같이 클래스의 메서드를 호출한다.
* 클래스의 <b>메서드 호출</b>

<pre>
student1.add_age(5)
</pre>

* 이때 student1 → self에 대응되고, 값 5 → offset에 대응된다.
* 클래스 <b>메서드 구현</b>

<pre>
def add_age(self, offset):
    self.age += offset
</pre>

````python
student1.add_age(10)
student1.show()
````

## 14.5.클래스 상속

- **상속**은 기존 클래스의 모든 기능을 물려받아 새로운 클래스를 만드는 방법이다.
- 기존 클래스(부모 클래스)의 코드를 그대로 사용할 수 있고, 필요하면 기능을 추가하거나 변경할 수도 있다.

**1) 상속의 기본 구조**
- 부모 클래스: 기존 기능을 정의한 클래스
- 자식 클래스: 부모 클래스의 기능을 물려받고, 추가 기능을 정의한 클래스

````python
# 기본 구조

# 부모 클래스
class Parents:
    def sing(self):
        print('sing a song')
    
# 자식 클래스
class Child(Parents):
    def dance(self):
        print('k-pop dance')

luckyboy = Child()

luckyboy.sing()
luckyboy.dance()
````

````python
# 기본 구조

# 부모 클래스
class Parents2:
    def sing(self):
        print('sing a song!')
    
# 자식 클래스
class Child2(Parents2):
    def sing(self):
        print('난 더더더잘해')

    # 오버라이딩?이 되는거네.
    def dance(self):
        print('k-pop dance')
    

luckyboy = Child2()

luckyboy.sing() # 부모 클래스의 메서드를 사용
luckyboy.dance() # 자식 클래스의 메서드를 사용

````

**4) 메서드 오버라이딩 (Method Overriding)**
- 자식 클래스에서 부모 클래스의 메서드를 재정의할 수 있다.
- 부모의 동작을 변경하거나 확장할 때 유용하다.

````python
class Parents:
    def sing(self):
        print('Old song')

class Child(Parents):
    def sing(self):
        print('New song')

child = Child()
child.sing()

Parents = Parents()
Parents.sing()
````

**5) 다중 상속**
- 여러 부모 클래스를 동시에 상속받을 수 있다.
- 하지만 코드가 복잡해질 수 있으니 주의해야 한다.

````python
class Parents1:
    def greet(self):
        print('Hello from Parent!')

class Parents2:
    def greet(self):
        print('Hello from parents2')

class Child(Parents1, Parents2):
    pass

child = Child()
child.greet()
````

````python
class Parents1:
    def greet(self):
        print('Hello from Parent!')

class Parents2:
    def greet(self):
        print('Hello from parents2')

# 순서에 따라 다르다. (child, 첫번째, 두번째)
class Child(Parents2, Parents1):
    pass

child = Child()
child.greet()
````
